```python
import sys
import random
import datetime
import pandas as pd
import matplotlib.pyplot as plt
import os
import numpy as np # Make sure numpy is imported

# Use perf_counter for high-resolution timing
from time import perf_counter

from ortools.constraint_solver import routing_enums_pb2, pywrapcp
# NO LONGER USING HAVERSINE FOR CORE DISTANCE CALCULATION
# from haversine import haversine, Unit 
import folium
import osmnx as ox
import networkx as nx # Need networkx for path length calculation

# For Jupyter display if needed
IN_JUPYTER = 'ipykernel' in sys.modules
if IN_JUPYTER:
    from IPython.display import display

# Configure OSMnx
ox.config(log_console=True, use_cache=True) # Enable caching

###############################################################################
# 1) SCENARIO DEFINITIONS (Case Studies)
###############################################################################
# We define three scenarios to reflect small, medium, and large region demands.
# Each scenario can be repeated or slightly changed if you want multiple runs.
TEST_CONFIGS = [
    # Case Study A: small region, fewer requests (Increase region size slightly for better node finding)
    {"num_requests": 10, "num_vehicles": 2, "region_size_deg": 0.08, "place_name": "Lake City, Florida, USA"}, # Use degrees for bbox
    # Case Study B: moderate region, moderate requests
    {"num_requests": 20, "num_vehicles": 4, "region_size_deg": 0.15, "place_name": "Lake City, Florida, USA"},
    # Case Study C: larger region, more requests
    {"num_requests": 30, "num_vehicles": 6, "region_size_deg": 0.25, "place_name": "Lake City, Florida, USA"},
]
TIME_LIMIT_SEC = 60  # Increase if needed, especially with network calcs
AVG_SPEED_MPH = 30 # Average speed for time calculation

###############################################################################
# 2) DATA CREATION (REAL ADDRESSES VIA OSMNX)
###############################################################################

def get_graph_and_nodes(place_name, region_size_deg):
    """
    Downloads/loads the street network graph for a given area and selects nodes.
    Returns the graph, a list of selected OSMnx node IDs, and the center coords.
    """
    # Geocode the center point
    try:
        center_point = ox.geocode(place_name)
        center_lat, center_lon = center_point[0], center_point[1]
        north = center_lat + region_size_deg / 2
        south = center_lat - region_size_deg / 2
        east = center_lon + region_size_deg / 2
        west = center_lon - region_size_deg / 2
        print(f"Getting graph for bbox: N={north:.4f}, S={south:.4f}, E={east:.4f}, W={west:.4f}")
        G = ox.graph_from_bbox(north, south, east, west, network_type="drive", simplify=True)
        # Ensure graph is strongly connected for routing, otherwise use largest component
        if not nx.is_strongly_connected(G):
            print("Graph is not strongly connected. Using the largest strongly connected component.")
            G = G.subgraph(max(nx.strongly_connected_components(G), key=len)).copy()

        if not G.nodes:
             raise ValueError("No nodes found in the specified graph area.")

        print(f"Graph loaded with {len(G.nodes)} nodes and {len(G.edges)} edges.")
        return G, center_lat, center_lon

    except Exception as e:
        print(f"Error getting graph for {place_name}: {e}")
        sys.exit(f"Could not obtain graph for {place_name}. Check place name and connectivity.")


def create_real_address_data(G, center_lat, center_lon, num_requests=8, num_vehicles=3):
    """
    Generates a dial-a-ride problem using actual road network nodes from OSMnx graph G.
    Calculates distances and times based on the network.
    """
    num_locations_needed = 1 + 2 * num_requests # Depot + N pickups + N dropoffs
    all_node_ids = list(G.nodes)

    if len(all_node_ids) < num_locations_needed:
        raise ValueError(f"Not enough nodes ({len(all_node_ids)}) in the graph for {num_locations_needed} required locations. Increase region size or use a denser area.")

    # --- Select Nodes ---
    # For depot, maybe find node closest to the original center? Or just random. Let's use random for now.
    # Ensure selected nodes are unique
    selected_node_ids = random.sample(all_node_ids, num_locations_needed)

    depot_node_id = selected_node_ids[0]
    request_node_ids = selected_node_ids[1:] # These will be shuffled for P/D pairs

    # --- Create Locations List and Mappings ---
    locations = [] # List of (name, lat, lon) tuples
    osmnx_node_ids_in_order = [] # Keep track of the OSMnx ID for each index in locations

    # Add Depot
    depot_lat = G.nodes[depot_node_id]['y']
    depot_lon = G.nodes[depot_node_id]['x']
    locations.append(("Depot", depot_lat, depot_lon))
    osmnx_node_ids_in_order.append(depot_node_id)

    # Add Pickup/Dropoff locations
    requests = []
    accessibility_requirements = {}
    node_index_counter = 1 # Start node index after depot (index 0)

    # Shuffle request nodes to create random P/D pairs
    random.shuffle(request_node_ids)

    for i in range(num_requests):
        pickup_node_id = request_node_ids[2*i]
        dropoff_node_id = request_node_ids[2*i + 1]

        # Ensure pickup and dropoff are not the same node (highly unlikely with random sampling, but good practice)
        if pickup_node_id == dropoff_node_id:
             print(f"Warning: Sampled same node for pickup/dropoff for Request {i+1}. Resampling dropoff...")
             available_nodes = list(set(all_node_ids) - set(selected_node_ids))
             if not available_nodes:
                 # This should ideally not happen if initial check passed
                 raise ValueError("Ran out of unique nodes during P/D assignment.")
             new_dropoff_node_id = random.choice(available_nodes)
             # Replace in selected_node_ids list to avoid duplicate selection later if needed
             selected_node_ids[1 + 2*i + 1] = new_dropoff_node_id
             dropoff_node_id = new_dropoff_node_id


        pickup_lat = G.nodes[pickup_node_id]['y']
        pickup_lon = G.nodes[pickup_node_id]['x']
        dropoff_lat = G.nodes[dropoff_node_id]['y']
        dropoff_lon = G.nodes[dropoff_node_id]['x']

        # Full-day time windows (can be customized later)
        window = (0, 1440) # 0 to 24*60 minutes

        needs_bike = (random.random() < 0.5)
        needs_wheelchair = (random.random() < 0.5)

        req = {
            "name": f"R{i+1}",
            "pickup_lat": pickup_lat, "pickup_lon": pickup_lon, "pickup_node_id": pickup_node_id,
            "dropoff_lat": dropoff_lat, "dropoff_lon": dropoff_lon, "dropoff_node_id": dropoff_node_id,
            "time_window": window,
            "needs_bike": needs_bike,
            "needs_wheelchair": needs_wheelchair
        }
        requests.append(req)

        # Add pickup location to list
        locations.append((f"{req['name']}_pickup", req["pickup_lat"], req["pickup_lon"]))
        osmnx_node_ids_in_order.append(pickup_node_id)
        pickup_loc_idx = node_index_counter
        node_index_counter += 1

        # Add dropoff location to list
        locations.append((f"{req['name']}_dropoff", req["dropoff_lat"], req["dropoff_lon"]))
        osmnx_node_ids_in_order.append(dropoff_node_id)
        dropoff_loc_idx = node_index_counter
        node_index_counter += 1

        # Store accessibility for pickup node index only (as per original logic)
        accessibility_requirements[pickup_loc_idx] = (req["needs_bike"], req["needs_wheelchair"])

    # --- Create OR-Tools Data Dictionary ---
    data = {}
    data["num_vehicles"] = num_vehicles
    data["depot"] = 0 # Index 0 in the locations list
    # Large capacity to avoid capacity constraints overshadowing the test
    data["vehicle_capacities"] = [9999] * num_vehicles # Kept large capacity

    num_locs = len(locations)
    print(f"Total locations including depot: {num_locs}")

    # --- Calculate Distance and Time Matrices using NetworkX on the Graph G ---
    dist_mat = np.full((num_locs, num_locs), -1.0) # Use float for meters first
    time_mat_minutes = np.full((num_locs, num_locs), -1.0)

    print("Calculating distance and time matrices using road network...")
    for i in range(num_locs):
        start_node = osmnx_node_ids_in_order[i]
        for j in range(num_locs):
            if i == j:
                dist_mat[i, j] = 0.0
                time_mat_minutes[i, j] = 0.0
                continue
            
            end_node = osmnx_node_ids_in_order[j]
            try:
                # Get distance in meters
                length_meters = nx.shortest_path_length(G, source=start_node, target=end_node, weight='length')
                dist_mat[i, j] = length_meters * 0.000621371 # Convert meters to miles

                # Calculate time based on distance and average speed
                time_mat_minutes[i, j] = (dist_mat[i, j] / AVG_SPEED_MPH) * 60.0 # Time in minutes

            except nx.NetworkXNoPath:
                print(f"Warning: No path found between node {start_node} (idx {i}) and node {end_node} (idx {j}). Setting large distance/time.")
                # Handle no path: Set a very large penalty distance/time
                dist_mat[i, j] = 999999.0 # Very large distance in miles
                time_mat_minutes[i, j] = 999999.0 # Very large time in minutes
            except Exception as e:
                 print(f"Error calculating path between {start_node} and {end_node}: {e}")
                 dist_mat[i, j] = 999999.0
                 time_mat_minutes[i, j] = 999999.0


    # Convert matrices to integer milliseconds/scaled miles for OR-Tools
    # Scaling factor helps maintain precision when converting to int
    # Original code used miles * 1000 and minutes * 1000
    data["distance_matrix"] = (dist_mat * 1000).astype(int).tolist()
    data["time_matrix"] = (time_mat_minutes * 1000).astype(int).tolist() # Time in milliseconds

    # Check if any -1 values remain (means error in calculation)
    if np.any(dist_mat < 0) or np.any(time_mat_minutes < 0):
         print("Warning: Negative values detected in distance/time matrix after calculation. Check for errors.")
         # Optional: add logic to handle this, e.g., replace negatives with large numbers
         data["distance_matrix"] = np.maximum(0, np.array(data["distance_matrix"])).tolist()
         data["time_matrix"] = np.maximum(0, np.array(data["time_matrix"])).tolist()


    # --- Setup Demands, Pickups/Deliveries, Time Windows ---
    data["demands"] = [0] * num_locs
    data["pickups_deliveries"] = []
    data["time_windows"] = [] # In minutes * 1000 for OR-Tools (scaled)

    # Depot time window (scaled)
    data["time_windows"].append((0 * 60000, 1440 * 60000)) # 0 to 24*60 minutes * 1000 (milliseconds, adjusted scale to match time matrix)

    passenger_node_to_name = {} # Map location index to request name
    for i, req in enumerate(requests):
        # Calculate indices based on the order they were added
        pu_idx = 1 + 2*i
        do_idx = pu_idx + 1
        data["pickups_deliveries"].append([pu_idx, do_idx])
        data["demands"][pu_idx] = 1 # Pickup adds 1 passenger
        data["demands"][do_idx] = -1 # Dropoff removes 1 passenger
        passenger_node_to_name[pu_idx] = req["name"]
        passenger_node_to_name[do_idx] = req["name"]

        # Add time windows (scaled to milliseconds)
        tw_start_ms = req["time_window"][0] * 60000 # minutes * 60 * 1000
        tw_end_ms = req["time_window"][1] * 60000   # minutes * 60 * 1000
        data["time_windows"].append((tw_start_ms, tw_end_ms)) # Pickup window
        data["time_windows"].append((tw_start_ms, tw_end_ms)) # Dropoff window

    # --- Vehicle Accessibility ---
    data["vehicles"] = []
    for v in range(num_vehicles):
        bike_capable = (random.random() < 0.5)
        wheelchair_accessible = (random.random() < 0.5)
        data["vehicles"].append({
            "bike_capable": bike_capable,
            "wheelchair_accessible": wheelchair_accessible
        })

    # Return the graph G and the ordered list of OSMnx node IDs along with other data
    return data, locations, passenger_node_to_name, accessibility_requirements, requests, G, osmnx_node_ids_in_order


###############################################################################
# 3) SOLVE ROUTING (Minor Adjustments for Time Scale)
###############################################################################
def solve_routing(data, locations, passenger_node_to_name, accessibility_requirements,
                  fs_strategy="PCA", ls_heuristic="GLS", time_limit=60):
    """
    Solves the routing problem with no skipping.
    Returns metrics and the route for each vehicle.
    NOTE: Time dimension units need to match the time_matrix units (milliseconds).
    """
    manager = pywrapcp.RoutingIndexManager(len(data["distance_matrix"]),
                                           data["num_vehicles"],
                                           data["depot"])
    routing = pywrapcp.RoutingModel(manager)

    # --- Distance Dimension ---
    def distance_callback(from_index, to_index):
        # Returns scaled distance (e.g., miles * 1000)
        from_node = manager.IndexToNode(from_index)
        to_node = manager.IndexToNode(to_index)
        return data["distance_matrix"][from_node][to_node]

    dist_cb_idx = routing.RegisterTransitCallback(distance_callback)
    routing.SetArcCostEvaluatorOfAllVehicles(dist_cb_idx)

    # --- Demand Dimension (Capacity) ---
    def demand_callback(from_index):
        # Returns passenger demand (+1 pickup, -1 dropoff)
        from_node = manager.IndexToNode(from_index)
        return data["demands"][from_node]

    demand_cb_idx = routing.RegisterUnaryTransitCallback(demand_callback)
    routing.AddDimensionWithVehicleCapacity(demand_cb_idx, 0, # No slack
                                            data["vehicle_capacities"],
                                            True, # Start cumul to zero
                                            "Capacity")

    # --- Time Dimension ---
    def time_callback(from_index, to_index):
        # Returns time in milliseconds
        from_node = manager.IndexToNode(from_index)
        to_node = manager.IndexToNode(to_index)
        return data["time_matrix"][from_node][to_node]

    time_cb_idx = routing.RegisterTransitCallback(time_callback)
    # Horizon needs to be large enough for scaled time (milliseconds)
    # 24 hours * 60 min/hr * 60 sec/min * 1000 ms/sec = 86,400,000 ms
    horizon = 86400000
    # Slack can be large if needed, but relates to waiting time. Set reasonably large.
    slack_max = 30 * 60000 # 30 minutes in ms
    routing.AddDimension(time_cb_idx,
                         slack_max, # Allowed waiting time (slack) at nodes in ms
                         horizon,   # Vehicle end time horizon in ms
                         False,     # Don't force start cumul to zero (depot has TW)
                         "Time")
    time_dim = routing.GetDimensionOrDie("Time")

    # Apply time windows (already in milliseconds from data creation)
    for loc_idx, tw in enumerate(data["time_windows"]):
        if loc_idx == data["depot"]: continue # Skip depot if start cumul=0 is false? Check OR-Tools behavior
        index = manager.NodeToIndex(loc_idx)
        time_dim.CumulVar(index).SetRange(tw[0], tw[1])

    # Add depot start/end time windows if needed (especially if start_cumul_to_zero=False)
    depot_idx = manager.NodeToIndex(data["depot"])
    for v in range(data["num_vehicles"]):
         routing.AddVariableMinimizedByFinalizer(time_dim.CumulVar(routing.Start(v)))
         routing.AddVariableMinimizedByFinalizer(time_dim.CumulVar(routing.End(v)))
         time_dim.CumulVar(routing.Start(v)).SetRange(data["time_windows"][data["depot"]][0], data["time_windows"][data["depot"]][1])
         # Allow ending anytime within the horizon
         time_dim.CumulVar(routing.End(v)).SetRange(data["time_windows"][data["depot"]][0], horizon)


    # --- Pickup and Delivery Constraints ---
    for (pu_idx, do_idx) in data["pickups_deliveries"]:
        pu_node_index = manager.NodeToIndex(pu_idx)
        do_node_index = manager.NodeToIndex(do_idx)
        routing.AddPickupAndDelivery(pu_node_index, do_node_index)
        # Ensure same vehicle serves pickup and dropoff
        routing.solver().Add(routing.VehicleVar(pu_node_index) == routing.VehicleVar(do_node_index))
        # Ensure pickup happens before dropoff in time
        routing.solver().Add(time_dim.CumulVar(pu_node_index) <= time_dim.CumulVar(do_node_index))

    # --- Accessibility Constraints ---
    for pu_loc_idx, (needs_bike, needs_wheelchair) in accessibility_requirements.items():
        # pu_loc_idx is the index in the `locations` list
        allowed_vehicles = []
        for v_idx, vehicle_spec in enumerate(data["vehicles"]):
            bike_ok = needs_bike and vehicle_spec["bike_capable"]
            wc_ok = needs_wheelchair and vehicle_spec["wheelchair_accessible"]
            # If no special needs, any vehicle is okay
            # If bike needed, vehicle must be bike capable
            # If wheelchair needed, vehicle must be wheelchair capable
            if not needs_bike and not needs_wheelchair:
                allowed_vehicles.append(v_idx)
            elif needs_bike and not needs_wheelchair and vehicle_spec["bike_capable"]:
                 allowed_vehicles.append(v_idx)
            elif not needs_bike and needs_wheelchair and vehicle_spec["wheelchair_accessible"]:
                 allowed_vehicles.append(v_idx)
            elif needs_bike and needs_wheelchair and vehicle_spec["bike_capable"] and vehicle_spec["wheelchair_accessible"]:
                 allowed_vehicles.append(v_idx)

        if not allowed_vehicles:
             print(f"Warning: No suitable vehicle found for request at location index {pu_loc_idx} (needs bike: {needs_bike}, needs wheelchair: {needs_wheelchair}). This request might be unserved or cause infeasibility.")
             # If no vehicles are suitable, technically the request cannot be served by any vehicle.
             # OR-Tools might handle this by making the solution infeasible if the node *must* be visited.
             # If skipping is allowed (not implemented here), it would be skipped.
             # For now, let's allow the solver to try, it might fail if no vehicle can take it.
             # You could explicitly make it unperformed using penalties if needed.
             # routing.SetAllowedVehiclesForIndex([], manager.NodeToIndex(pu_loc_idx)) # This would likely make it infeasible
        else:
             # print(f"Setting allowed vehicles for index {pu_loc_idx}: {allowed_vehicles}")
             pu_node_index = manager.NodeToIndex(pu_loc_idx)
             routing.SetAllowedVehiclesForIndex(allowed_vehicles, pu_node_index)
             # Also constrain the dropoff? The P&D constraint should enforce same vehicle anyway.


    # --- Search Parameters ---
    search_params = pywrapcp.DefaultRoutingSearchParameters()
    # Strategies like PATH_CHEAPEST_ARC or SAVINGS might be good starting points
    search_params.first_solution_strategy = (
        # routing_enums_pb2.FirstSolutionStrategy.AUTOMATIC
         routing_enums_pb2.FirstSolutionStrategy.PATH_CHEAPEST_ARC
        # routing_enums_pb2.FirstSolutionStrategy.SAVINGS
    )
    # Metaheuristics like GUIDED_LOCAL_SEARCH or TABU_SEARCH help improve
    search_params.local_search_metaheuristic = (
        routing_enums_pb2.LocalSearchMetaheuristic.GUIDED_LOCAL_SEARCH
        # routing_enums_pb2.LocalSearchMetaheuristic.TABU_SEARCH
        # routing_enums_pb2.LocalSearchMetaheuristic.AUTOMATIC
    )
    search_params.time_limit.FromSeconds(time_limit)
    search_params.log_search = True # Log progress

    # --- Solve ---
    print("Starting solver...")
    solution = routing.SolveWithParameters(search_params)

    # --- Process Solution ---
    if not solution:
        print("No solution found!")
        # You might want to inspect routing.status() for reasons
        status_map = {0: "ROUTING_NOT_SOLVED", 1: "ROUTING_SUCCESS", 2: "ROUTING_FAIL", 3: "ROUTING_FAIL_TIMEOUT", 4: "ROUTING_INVALID"}
        print(f"Solver status: {routing.status()} ({status_map.get(routing.status(), 'UNKNOWN')})")
        return None, None

    print("Solution found!")

    # Extract routes
    routes = [] # List of lists, each sublist contains node indices for a vehicle route
    for v in range(data["num_vehicles"]):
        index = routing.Start(v)
        route_nodes = []
        while not routing.IsEnd(index):
            node_idx = manager.IndexToNode(index)
            route_nodes.append(node_idx)
            index = solution.Value(routing.NextVar(index))
        # Add the end node index
        route_nodes.append(manager.IndexToNode(routing.End(v)))
        # Only add non-empty routes (vehicle might not be used)
        # A route just containing Depot Start -> Depot End might have length 2 if depot is start/end node
        if len(route_nodes) > 2 or (len(route_nodes) == 2 and route_nodes[0] != route_nodes[1]):
             routes.append(route_nodes)
        else:
             print(f"Vehicle {v} not used.")
             # Add an empty route placeholder if you need the list size to match num_vehicles
             # routes.append([])


    # --- Calculate Metrics ---
    total_dist_miles = solution.ObjectiveValue() / 1000.0 # Objective is scaled distance

    # Calculate total time from solution (sum of route times)
    total_time_ms = 0
    for v in range(data["num_vehicles"]):
         # end_time = solution.Min(time_dim.CumulVar(routing.End(v))) # This gives earliest end time
         # start_time = solution.Min(time_dim.CumulVar(routing.Start(v))) # This gives earliest start time
         # total_time_ms += (end_time - start_time) # This is duration, not necessarily travel time

         # Let's recalculate time from the route itself for consistency check
         vehicle_route = routes[v] if v < len(routes) else []
         route_time_ms = 0
         if len(vehicle_route) > 1:
             for i in range(len(vehicle_route) - 1):
                 from_node = vehicle_route[i]
                 to_node = vehicle_route[i+1]
                 # Lookup time from the original time matrix (in ms)
                 route_time_ms += data["time_matrix"][from_node][to_node]
         total_time_ms += route_time_ms

    sim_time_min = total_time_ms / 60000.0 # Convert total ms to minutes


    # Gather vehicle details
    vehicle_details = []
    route_segment_distances = [] # In miles
    assigned_routes = routes # Use the extracted routes

    for v_idx, route in enumerate(assigned_routes): # Iterate through actual used routes
        route_dist_miles = 0.0
        non_depot_stops = 0
        load_over_route = []
        current_load = 0
        # Get start/end times for the vehicle route from solution
        # start_time_ms = solution.Min(time_dim.CumulVar(routing.Start(v_idx)))
        # end_time_ms = solution.Min(time_dim.CumulVar(routing.End(v_idx)))

        if len(route) > 1: # Need at least two points for a segment
            # Initial load at the start (usually depot, load 0)
            start_node_idx = route[0]
            current_load += data["demands"][start_node_idx]
            load_over_route.append(current_load)
            if start_node_idx != data["depot"]: non_depot_stops += 1 # Count if start isn't depot

            for i in range(len(route)-1):
                from_node_idx = route[i]
                to_node_idx = route[i+1]

                # Use distance matrix for segment length (miles * 1000)
                dist_leg_scaled = data["distance_matrix"][from_node_idx][to_node_idx]
                dist_leg_miles = dist_leg_scaled / 1000.0
                route_dist_miles += dist_leg_miles
                route_segment_distances.append(dist_leg_miles)

                # Update load based on the *destination* node of the segment (to_node_idx)
                current_load += data["demands"][to_node_idx]
                load_over_route.append(current_load)

                # Count non-depot stops
                if to_node_idx != data["depot"]:
                    non_depot_stops += 1

            # Correcting stops count: Original code counted `from_node`, let's count unique non-depot nodes visited.
            unique_non_depot_nodes = set(r for r in route if r != data["depot"])
            non_depot_stops = len(unique_non_depot_nodes)


        vehicle_details.append({
            "vehicle_id": v_idx, # Use the index from the assigned_routes list
            "route_distance": route_dist_miles,
            "stops": non_depot_stops,
            "load_list": load_over_route
            # "start_time_min": start_time_ms / 60000.0,
            # "end_time_min": end_time_ms / 60000.0,
        })

    # Count accessibility requests/vehicles
    bike_requests = sum(1 for (b, w) in accessibility_requirements.values() if b)
    wheelchair_requests = sum(1 for (b, w) in accessibility_requirements.values() if w)
    bike_vehicles = sum(1 for v in data["vehicles"] if v["bike_capable"])
    wheelchair_vehicles = sum(1 for v in data["vehicles"] if v["wheelchair_accessible"])

    metrics = {
        "distance_miles": total_dist_miles,
        "sim_time_min": sim_time_min, # Total travel time in minutes
        "vehicle_details": vehicle_details,
        "segment_distances": route_segment_distances, # List of segment lengths in miles
        "bike_requests": bike_requests,
        "wheelchair_requests": wheelchair_requests,
        "bike_vehicles": bike_vehicles,
        "wheelchair_vehicles": wheelchair_vehicles
    }
    return metrics, assigned_routes # Return only the routes for vehicles that are used


###############################################################################
# 4) VISUALIZATION (MAP - Enhanced with Network Paths)
###############################################################################
def visualize_routes(routes, data, locations, passenger_node_to_name,
                     accessibility_requirements, vehicle_colors,
                     G, osmnx_node_ids_in_order): # Pass G and node IDs
    """
    Creates a Folium map using actual road network paths between stops.
    """
    from folium.features import DivIcon
    if not routes:
        print("No routes to visualize.")
        return folium.Map(location=[30.1897, -82.6393], zoom_start=10) # Default map

    # Use depot location from data
    depot_loc_idx = data["depot"]
    depot_lat, depot_lon = locations[depot_loc_idx][1], locations[depot_loc_idx][2]
    m = folium.Map(location=[depot_lat, depot_lon], zoom_start=12) # Adjust zoom as needed

    # Check if graph G is provided
    if G is None:
        print("Warning: No graph G provided for visualization. Drawing straight lines.")

    # Create a mapping from location index back to OSMnx node ID
    index_to_osmnx_id = {i: node_id for i, node_id in enumerate(osmnx_node_ids_in_order)}

    for v_idx, route in enumerate(routes): # Routes only contains used vehicles
        if not route: continue # Skip if route is empty
        color = vehicle_colors[v_idx % len(vehicle_colors)]

        # Plot route segments using network paths if G is available
        if G and len(route) > 1:
            for i in range(len(route) - 1):
                start_loc_idx = route[i]
                end_loc_idx = route[i+1]

                start_node_id = index_to_osmnx_id.get(start_loc_idx)
                end_node_id = index_to_osmnx_id.get(end_loc_idx)

                if start_node_id is None or end_node_id is None:
                     print(f"Warning: Missing OSMnx node ID for location index {start_loc_idx} or {end_loc_idx}. Skipping segment.")
                     continue

                if start_node_id == end_node_id: continue # Skip self-loops

                try:
                    # Get the sequence of OSMnx node IDs for the shortest path
                    path_node_ids = nx.shortest_path(G, source=start_node_id, target=end_node_id, weight='length')
                    if len(path_node_ids) > 1:
                         # Get coordinates for each node in the path
                         path_coords = [(G.nodes[node_id]['y'], G.nodes[node_id]['x']) for node_id in path_node_ids]
                         # Draw the path segment
                         folium.PolyLine(path_coords, color=color, weight=3, opacity=0.7).add_to(m)
                    elif len(path_node_ids) == 1: # Start and end are the same - draw a point?
                         pass # Or handle as needed

                except nx.NetworkXNoPath:
                    print(f"Warning: No path found between {start_node_id} and {end_node_id} for visualization. Drawing straight line.")
                    # Fallback: draw straight line
                    start_coords = (locations[start_loc_idx][1], locations[start_loc_idx][2])
                    end_coords = (locations[end_loc_idx][1], locations[end_loc_idx][2])
                    folium.PolyLine([start_coords, end_coords], color=color, weight=2, opacity=0.5, dash_array='5, 5').add_to(m)
                except Exception as e:
                    print(f"Error getting path for visualization between {start_node_id} and {end_node_id}: {e}")
                    # Fallback: draw straight line
                    start_coords = (locations[start_loc_idx][1], locations[start_loc_idx][2])
                    end_coords = (locations[end_loc_idx][1], locations[end_loc_idx][2])
                    folium.PolyLine([start_coords, end_coords], color=color, weight=2, opacity=0.5, dash_array='5, 5').add_to(m)

        # Add markers with labels for each stop
        for stop_order, loc_idx in enumerate(route, start=1):
            lat, lon = locations[loc_idx][1], locations[loc_idx][2]
            node_name = locations[loc_idx][0] # e.g., "Depot", "R1_pickup"

            if loc_idx == data["depot"]:
                point_type = "Depot"
                request_name = "Depot"
                marker_color = 'black'
            else:
                # Determine if pickup or dropoff based on demand
                if data["demands"][loc_idx] == 1:
                    point_type = "Pickup"
                elif data["demands"][loc_idx] == -1:
                    point_type = "Dropoff"
                else: # Should not happen for non-depot nodes in P&D
                    point_type = "Stop"
                request_name = passenger_node_to_name.get(loc_idx, "?")
                marker_color = color # Use vehicle color for P/D stops


            # Check accessibility needs (stored against pickup index)
            # Need to find the corresponding pickup index if this is a dropoff
            is_pickup = data["demands"][loc_idx] == 1
            accessibility_index = loc_idx if is_pickup else -1
            if not is_pickup and loc_idx != data["depot"]:
                # Find the matching pickup index from pickups_deliveries list
                for p_idx, d_idx in data["pickups_deliveries"]:
                    if d_idx == loc_idx:
                        accessibility_index = p_idx
                        break

            needs_bike, needs_wheelchair = accessibility_requirements.get(accessibility_index, (False, False))

            icon_str = ""
            if needs_bike: icon_str += " 🚲"
            if needs_wheelchair: icon_str += " ♿"

            label_text = f"V{v_idx}, S{stop_order}: {point_type} {request_name}{icon_str}"
            # Use standard markers first, then consider DivIcon if needed
            folium.Marker(
                location=[lat, lon],
                popup=label_text,
                tooltip=label_text,
                icon=folium.Icon(color=marker_color, icon='info-sign' if loc_idx != data["depot"] else 'home')
            ).add_to(m)

            # Optional: Keep DivIcon if preferred for always-visible labels
            # offset_lat = 0.0001 * (v_idx % 3 - 1) # Slight offset to avoid overlap
            # offset_lon = 0.0001 * (stop_order % 3 - 1)
            # folium.map.Marker(
            #     location=(lat + offset_lat, lon + offset_lon),
            #     icon=DivIcon(
            #         icon_size=(150, 36), icon_anchor=(0, 0),
            #         html=f'<div style="font-size: 8pt; font-weight: bold; color: white; background-color: {color}; border: 1px solid black; border-radius: 3px; padding: 2px; opacity: 0.8;">{label_text}</div>'
            #     )
            # ).add_to(m)

    return m


###############################################################################
# 5) BUILD RESULTS PAGE (HTML) - No changes needed
###############################################################################
def build_results_page(df, map_files, chart_file):
    html = "<html><head><title>Routing Results</title></head><body>"
    html += "<h1>Routing Test Results (Real Addresses)</h1>" # Updated Title
    html += "<p>Below are the aggregated metrics from our three case studies, each representing a distinct scenario in terms of number of requests, vehicles, and region size, using real road network nodes and distances.</p>" # Updated Desc
    html += "<h2>Aggregated Metrics</h2>"
    html += df.to_html(index=False)
    html += f"<h2>Comparison Chart (Basic)</h2><p>This bar chart compares total network distance (left) and simulation time (right) across the scenarios. We see how the system scales with the number of requests and region size using road network calculations.</p>" # Updated Desc
    html += f"<img src='{chart_file}' alt='Grouped Bar Chart' style='max-width:800px;'>"
    html += "<h2>Route Maps</h2><ul>"
    for cfg, mapfile in map_files.items():
        html += f"<li><a href='{mapfile}' target='_blank'>Config: {cfg}</a> - View the Folium map for this scenario (routes follow roads).</li>" # Updated Desc
    html += "</ul></body></html>"
    with open("results_real_addresses.html", "w") as f: # Changed filename
        f.write(html)
    print("Results page saved to results_real_addresses.html")


###############################################################################
# 6) ANALYZE AND PLOT (CHARTS WITH DESCRIPTIONS) - Update Region Size Label
###############################################################################
def analyze_and_plot(all_runs_data):
    """
    Breaks down each scenario (case study) individually:
      - Distance per vehicle
      - Stops per vehicle
      - Accessibility demand vs. supply
      - Distribution of route segment distances
      - Load over route order
    Then prints a short overall conclusion.
    """
    import numpy as np # Ensure numpy is available

    # Print a general introduction for the analytics
    print("\n================= ANALYTICS & CASE STUDIES (Real Addresses) =================")
    print("Below we examine each case study (scenario) individually using real road network data.\n")

    # Sort the scenarios for consistent ordering (optional)
    all_runs_data = sorted(all_runs_data, key=lambda x: x["Config"])

    for idx, scenario_data in enumerate(all_runs_data, start=1):
        # Identify scenario
        scenario_label = scenario_data["Config"]
        case_study_title = f"Case Study #{idx}: {scenario_label}"

        # Extract relevant data
        vehicle_details = scenario_data["vehicle_details"]
        if not vehicle_details: # Handle cases where no vehicles were used / no solution
             print(f"\n--- {case_study_title} ---")
             print("No vehicle details available (likely no solution found or no routes assigned). Skipping plots for this case.")
             continue

        v_dists = [v["route_distance"] for v in vehicle_details]
        v_stops = [v["stops"] for v in vehicle_details]
        # Filter out zero distances if any vehicle wasn't used but still listed
        segment_dists = [d for d in scenario_data.get("segment_distances", []) if d > 0.01] # Use get w/ default, filter small vals

        bike_req = scenario_data["bike_requests"]
        wc_req = scenario_data["wheelchair_requests"]
        bike_veh = scenario_data["bike_vehicles"]
        wc_veh = scenario_data["wheelchair_vehicles"]

        # Print scenario summary
        print(f"\n--- {case_study_title} ---")
        print(f"Place Name: {scenario_data['PlaceName']}") # Added PlaceName
        print(f"Number of Requests: {scenario_data['NumRequests']}")
        print(f"Number of Vehicles: {scenario_data['NumVehicles']}")
        print(f"Region Size: ±{scenario_data['RegionSizeDeg']/2:.3f} degrees approx") # Updated Region Size Label
        print(f"Total Distance (miles): {scenario_data['distance_miles']:.2f}")
        print(f"Total Simulation Time (minutes): {scenario_data['sim_time_min']:.2f}")
        print(f"Bike Requests: {bike_req}, Wheelchair Requests: {wc_req}")
        print(f"Bike Vehicles: {bike_veh}, Wheelchair Vehicles: {wc_veh}")

        # --- Plotting --- (Handle cases with no data)

        # 1) Distance per Vehicle
        if v_dists:
            fig, ax = plt.subplots(figsize=(6,4))
            ax.bar(range(len(v_dists)), v_dists, color='skyblue')
            ax.set_title(f"{case_study_title}\nDistance per Vehicle")
            ax.set_xlabel("Vehicle ID (Used Vehicles)")
            ax.set_ylabel("Distance (miles)")
            ax.set_xticks(range(len(v_dists))) # Ensure ticks match number of bars
            ax.set_xticklabels([vd["vehicle_id"] for vd in vehicle_details]) # Use actual vehicle IDs
            plt.tight_layout()
            plt.show()
            print("Chart Explanation: Shows how total route distance (based on road network) is distributed among the used vehicles.\n")
        else: print("Skipping Distance per Vehicle plot (no data).")


        # 2) Stops per Vehicle
        if v_stops:
            fig, ax = plt.subplots(figsize=(6,4))
            ax.bar(range(len(v_stops)), v_stops, color='orange')
            ax.set_title(f"{case_study_title}\nNon-Depot Stops per Vehicle")
            ax.set_xlabel("Vehicle ID (Used Vehicles)")
            ax.set_ylabel("Number of unique non-depot stops")
            ax.set_xticks(range(len(v_stops))) # Ensure ticks match number of bars
            ax.set_xticklabels([vd["vehicle_id"] for vd in vehicle_details]) # Use actual vehicle IDs
            plt.tight_layout()
            plt.show()
            print("Chart Explanation: Shows how many unique pickups/dropoffs each used vehicle performed.\n")
        else: print("Skipping Stops per Vehicle plot (no data).")


        # 3) Accessibility Demand vs. Supply (Single Scenario)
        fig, ax = plt.subplots(figsize=(6,4))
        bar_width = 0.35
        r1 = np.arange(2) # Positions for Bike, Wheelchair groups
        r2 = [x + bar_width for x in r1]
        ax.bar(r1[0], bike_req, color='green', width=bar_width, edgecolor='grey', label='Bike Requests')
        ax.bar(r1[1], wc_req, color='red', width=bar_width, edgecolor='grey', label='Wheelchair Requests')
        ax.bar(r2[0], bike_veh, color='blue', width=bar_width, edgecolor='grey', label='Bike Vehicles')
        ax.bar(r2[1], wc_veh, color='purple', width=bar_width, edgecolor='grey', label='Wheelchair Vehicles')
        ax.set_xlabel('Accessibility Need', fontweight='bold')
        ax.set_xticks([r + bar_width/2 for r in range(2)])
        ax.set_xticklabels(['Bike', 'Wheelchair'])
        ax.set_ylabel("Count")
        ax.set_title(f"{case_study_title}\nAccessibility Demand vs. Supply")
        ax.legend()
        plt.tight_layout()
        plt.show()
        print("Chart Explanation: Compares the number of bike/wheelchair requests to the number of\n"
              "vehicles equipped for each. If requests exceed available vehicles, some demands may go unmet.\n")

        # 4) Distribution of Route Segment Distances
        if segment_dists:
            fig, ax = plt.subplots(figsize=(6,4))
            ax.hist(segment_dists, bins=10, color='gray', edgecolor='black') # Increased bins
            ax.set_title(f"{case_study_title}\nDistribution of Route Segment Distances")
            ax.set_xlabel("Segment Distance (miles, network)")
            ax.set_ylabel("Frequency")
            plt.tight_layout()
            plt.show()
            print("Chart Explanation: Shows histogram of individual road network distances between consecutive stops.\nReflects actual travel legs.\n")
        else: print("Skipping Segment Distance Distribution plot (no data).")


        # 5) Load Over Route Order
        loads_exist = any(v.get("load_list") for v in vehicle_details)
        if loads_exist:
            fig, ax = plt.subplots(figsize=(8,5)) # Slightly larger
            for vinfo in vehicle_details:
                v_id = vinfo["vehicle_id"]
                load_list = vinfo.get("load_list", [])
                if load_list: # Only plot if load list is not empty
                     ax.plot(range(len(load_list)), load_list, marker='o', linestyle='-', markersize=4, label=f"Vehicle {v_id}")
            ax.set_title(f"{case_study_title}\nPassengers Onboard Over Route Stops")
            ax.set_xlabel("Stop Index in Vehicle's Route")
            ax.set_ylabel("Passengers Onboard")
            ax.legend(fontsize='small')
            ax.grid(True, linestyle='--', alpha=0.6)
            plt.tight_layout()
            plt.show()
            print("Chart Explanation: Depicts how each vehicle's passenger load changes stop-by-stop along its calculated route.\n")
        else: print("Skipping Load Over Route Order plot (no data).")


    # Final Conclusion
    print("\n================= CONCLUSION (Real Addresses) =================")
    print("Using real road network nodes and distances provides a more realistic simulation.\n"
          "We observed how the OR-Tools solver planned routes based on actual street layouts.\n"
          "Total distance and time reflect network travel rather than straight lines.\n"
          "Challenges include ensuring sufficient network nodes in the selected area and handling potential 'no path' errors between nodes.\n"
          "Accessibility constraints were applied, potentially limiting solutions if demand exceeded supply for specific vehicle types.\n"
          "Overall, this approach offers higher fidelity for analyzing rural transit routing scenarios.\n")


###############################################################################
# 7) MAIN: RUN CASE STUDIES, PLOT, & GENERATE RESULTS
###############################################################################
def main():
    all_runs = []
    map_files = {}

    # Pre-fetch graphs for all unique places to potentially save time if places repeat
    graphs = {}
    center_coords = {}
    for config in TEST_CONFIGS:
        place = config["place_name"]
        if place not in graphs:
             print(f"--- Pre-loading graph for {place} ---")
             try:
                 # Note: region_size_deg might differ per config even for same place,
                 # potentially leading to different graphs.
                 # For simplicity here, we fetch based on place name only, assuming
                 # subsequent fetches for the same place might hit cache or be fast.
                 # A more robust approach might key the cache by place+size.
                 G, lat, lon = get_graph_and_nodes(place, config["region_size_deg"])
                 graphs[place] = G
                 center_coords[place] = (lat, lon)
             except Exception as e:
                 print(f"Fatal Error: Could not load graph for {place}. Exiting. Error: {e}")
                 return # Exit if essential graph loading fails

    # --- Main Loop ---
    for config in TEST_CONFIGS:
        num_requests = config["num_requests"]
        num_vehicles = config["num_vehicles"]
        region_size_deg = config["region_size_deg"]
        place_name = config["place_name"]
        # Use a slightly more descriptive label including place snippet
        place_label = "".join(filter(str.isalnum, place_name.split(",")[0]))[:10]
        config_label = f"{place_label}_Req{num_requests}_Veh{num_vehicles}_Reg{region_size_deg*100:.0f}" # Reg as degrees*100

        print(f"\n=== Running test config: {config_label} ===")
        print(f"   Place: {place_name}, Requests: {num_requests}, Vehicles: {num_vehicles}, Region Size (deg): {region_size_deg}")

        # Get the pre-fetched graph and center coords
        G = graphs.get(place_name)
        center_lat, center_lon = center_coords.get(place_name, (None, None))

        if G is None or center_lat is None:
             print(f"Error: Graph or center coordinates not found for {place_name}. Skipping config.")
             continue

        try:
            # Create data using the graph
            data, locations, passenger_node_to_name, accessibility_requirements, requests, G_used, osmnx_node_ids_in_order = \
                create_real_address_data(G=G, # Pass the graph
                                          center_lat=center_lat, # Pass center coords
                                          center_lon=center_lon,
                                          num_requests=num_requests,
                                          num_vehicles=num_vehicles)
            # Note: G_used should be the same as G here unless modified inside create_real_address_data

            # Solve routing
            metrics, routes = solve_routing(data, locations, passenger_node_to_name,
                                            accessibility_requirements,
                                            # fs_strategy="PCA", ls_heuristic="GLS", # Keep original or adjust
                                            time_limit=TIME_LIMIT_SEC)

        except ValueError as ve:
             print(f"Data creation error for config {config_label}: {ve}. Skipping.")
             continue # Skip this config if data creation fails (e.g., not enough nodes)
        except Exception as e:
             print(f"An unexpected error occurred during data creation or solving for {config_label}: {e}")
             import traceback
             traceback.print_exc()
             continue # Skip this config on other errors


        if metrics is None or routes is None:
            print(f"Solver failed for config {config_label}. Skipping results storage.")
            # Store failure information if desired
            all_runs.append({
                "Config": config_label,
                "NumRequests": num_requests,
                "NumVehicles": num_vehicles,
                "RegionSizeDeg": region_size_deg,
                "PlaceName": place_name,
                "distance_miles": -1, # Indicate failure
                "sim_time_min": -1,
                "vehicle_details": [],
                "segment_distances": [],
                "bike_requests": -1,
                "wheelchair_requests": -1,
                "bike_vehicles": -1,
                "wheelchair_vehicles": -1,
                "requests": [] # Or store the generated requests
            })
            continue

        # Store scenario-level info
        metrics.update({
            "Config": config_label,
            "NumRequests": num_requests,
            "NumVehicles": num_vehicles,
            "RegionSizeDeg": region_size_deg, # Store the degree size used
            "PlaceName": place_name, # Store place name
            "requests": requests # Store request details if needed later
        })
        all_runs.append(metrics)

        # Create a map for each scenario
        ts = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")
        mapfile = f"routes_{config_label}_{ts}.html"
        vehicle_colors = ['red', 'blue', 'green', 'purple', 'orange', 'darkred', 'lightblue', 'darkgreen', 'cadetblue', 'darkpurple'] # More colors
        try:
            my_map = visualize_routes(routes, data, locations, passenger_node_to_name,
                                      accessibility_requirements, vehicle_colors,
                                      G_used, osmnx_node_ids_in_order) # Pass G and node IDs
            my_map.save(mapfile)
            map_files[config_label] = mapfile
            print(f"Map saved to {mapfile}")

            # Display inline if in Jupyter
            if IN_JUPYTER:
                display(my_map)
        except Exception as e:
             print(f"Error during map visualization for {config_label}: {e}")
             import traceback
             traceback.print_exc()
             map_files[config_label] = "Error_creating_map"


    # --- Post-processing ---
    if not all_runs:
        print("No solutions found or data generated for any configuration!")
        return

    # Filter out failed runs for DataFrame and plotting if needed
    successful_runs = [r for r in all_runs if r["distance_miles"] >= 0]

    if not successful_runs:
        print("No successful solutions found across all configurations!")
        # Optionally print the details of the failed runs
        print("\n--- Failed Run Summary ---")
        df_failed = pd.DataFrame(all_runs)
        print(df_failed[['Config', 'NumRequests', 'NumVehicles', 'PlaceName', 'RegionSizeDeg']])
        return


    # Build a DataFrame of the main metrics for successful scenarios
    df = pd.DataFrame([
        {
            "Config": r["Config"],
            "Place": r["PlaceName"],
            "NumRequests": r["NumRequests"],
            "NumVehicles": r["NumVehicles"],
            "RegionSizeDeg": r["RegionSizeDeg"],
            "distance_miles": r["distance_miles"],
            "sim_time_min": r["sim_time_min"]
        } for r in successful_runs
    ])

    print("\n--- Results Summary (Successful Configurations) ---")
    print(df)

    # Basic bar charts for total distance & sim time
    df_indexed = df.set_index("Config")
    if not df_indexed.empty:
        try:
            fig, axes = plt.subplots(1, 2, figsize=(14, 7)) # Adjusted size for potentially longer labels
            df_indexed["distance_miles"].plot(kind="bar", ax=axes[0], rot=60, width=0.7) # Increased rotation, width
            axes[0].set_title("Total Network Distance (miles)")
            axes[0].set_ylabel("Miles")
            axes[0].tick_params(axis='x', labelsize=8) # Smaller x-axis labels if needed

            df_indexed["sim_time_min"].plot(kind="bar", ax=axes[1], rot=60, width=0.7) # Increased rotation, width
            axes[1].set_title("Total Simulation Time (min)")
            axes[1].set_ylabel("Minutes")
            axes[1].tick_params(axis='x', labelsize=8) # Smaller x-axis labels if needed

            plt.tight_layout() # Adjust layout
            chart_filename = "grouped_comparison_real_addresses.png" # New filename
            plt.savefig(chart_filename)
            plt.show()
            print(f"Basic grouped bar chart saved to {chart_filename}")

            # Build HTML results page
            build_results_page(df, map_files, chart_filename)

            # Create additional analysis & charts with textual explanations
            analyze_and_plot(successful_runs) # Analyze only successful runs

        except Exception as e:
            print(f"Error during plotting or HTML generation: {e}")
            import traceback
            traceback.print_exc()
    else:
        print("DataFrame is empty, skipping plotting and HTML generation.")


if __name__ == "__main__":
    main()
```
