```python
import sys
import random
import datetime
import pandas as pd
import matplotlib.pyplot as plt
import os

# Use perf_counter for high-resolution timing
from time import perf_counter

from ortools.constraint_solver import routing_enums_pb2, pywrapcp
from haversine import haversine, Unit
import folium
import osmnx as ox
import networkx as nx

# For Jupyter display if needed
IN_JUPYTER = 'ipykernel' in sys.modules
if IN_JUPYTER:
    from IPython.display import display

###############################################################################
# 1) SCENARIO DEFINITIONS (Case Studies)
###############################################################################
# We define three scenarios to reflect small, medium, and large region demands.
# Each scenario can be repeated or slightly changed if you want multiple runs.
TEST_CONFIGS = [
    # Case Study A: small region, fewer requests
    {"num_requests": 10, "num_vehicles": 2, "region_size": 0.05},
    # Case Study B: moderate region, moderate requests
    {"num_requests": 20, "num_vehicles": 4, "region_size": 0.1},
    # Case Study C: larger region, more requests
    {"num_requests": 30, "num_vehicles": 6, "region_size": 0.2},
]
TIME_LIMIT_SEC = 30  # Increase if needed

###############################################################################
# 2) DATA CREATION (RANDOM)
###############################################################################
def create_rural_data(num_requests=8, num_vehicles=3, region_size=0.1):
    """
    Generates a dial-a-ride problem with wide time windows and large capacities.
    Randomness ensures each run is unique. We do not fix seeds here.
    """
    import numpy as np
    center_lat, center_lon = 30.1897, -82.6393
    lat_range = region_size
    lon_range = region_size

    locations = [("Depot", center_lat, center_lon)]
    requests = []
    accessibility_requirements = {}

    for i in range(num_requests):
        pickup_lat = center_lat + random.uniform(-lat_range, lat_range)
        pickup_lon = center_lon + random.uniform(-lon_range, lon_range)
        dropoff_lat = center_lat + random.uniform(-lat_range, lat_range)
        dropoff_lon = center_lon + random.uniform(-lon_range, lon_range)

        # Full-day time windows
        window = (0, 1440)

        needs_bike = (random.random() < 0.5)
        needs_wheelchair = (random.random() < 0.5)
        
        req = {
            "name": f"R{i+1}",
            "pickup_lat": pickup_lat,
            "pickup_lon": pickup_lon,
            "dropoff_lat": dropoff_lat,
            "dropoff_lon": dropoff_lon,
            "time_window": window,
            "needs_bike": needs_bike,
            "needs_wheelchair": needs_wheelchair
        }
        requests.append(req)
    
    # Build location list
    for i, req in enumerate(requests):
        # pickup
        locations.append((f"{req['name']}_pickup", req["pickup_lat"], req["pickup_lon"]))
        # dropoff
        locations.append((f"{req['name']}_dropoff", req["dropoff_lat"], req["dropoff_lon"]))
        # Store accessibility for pickup node only
        accessibility_requirements[1 + 2*i] = (req["needs_bike"], req["needs_wheelchair"])
    
    data = {}
    data["num_vehicles"] = num_vehicles
    data["depot"] = 0
    # Large capacity to avoid capacity constraints overshadowing the test
    data["vehicle_capacities"] = [9999] * num_vehicles

    num_locs = len(locations)
    dist_mat = np.zeros((num_locs, num_locs))
    for i in range(num_locs):
        for j in range(num_locs):
            coord1 = (locations[i][1], locations[i][2])
            coord2 = (locations[j][1], locations[j][2])
            dist_mat[i][j] = haversine(coord1, coord2, unit=Unit.MILES)
    
    data["distance_matrix"] = (dist_mat * 1000).astype(int).tolist()
    # Time matrix based on 30 mph
    time_mat = (dist_mat / 30.0) * 60.0
    data["time_matrix"] = (time_mat * 1000).astype(int).tolist()
    
    data["demands"] = [0] * num_locs
    data["pickups_deliveries"] = []
    data["time_windows"] = []
    
    # Depot time window
    data["time_windows"].append((0, 1440))
    
    passenger_node_to_name = {}
    for i, req in enumerate(requests):
        pu_idx = 1 + 2*i
        do_idx = pu_idx + 1
        data["pickups_deliveries"].append([pu_idx, do_idx])
        data["demands"][pu_idx] = 1
        data["demands"][do_idx] = -1
        passenger_node_to_name[pu_idx] = req["name"]
        passenger_node_to_name[do_idx] = req["name"]
        data["time_windows"].append(req["time_window"])
        data["time_windows"].append(req["time_window"])
    
    # Vehicle accessibility with 50% chance each
    data["vehicles"] = []
    for v in range(num_vehicles):
        bike_capable = (random.random() < 0.5)
        wheelchair_accessible = (random.random() < 0.5)
        data["vehicles"].append({
            "bike_capable": bike_capable,
            "wheelchair_accessible": wheelchair_accessible
        })
    
    return data, locations, passenger_node_to_name, accessibility_requirements, requests

###############################################################################
# 3) SOLVE ROUTING
###############################################################################
def solve_routing(data, locations, passenger_node_to_name, accessibility_requirements,
                  fs_strategy="PCA", ls_heuristic="GLS", time_limit=60):
    """
    Solves the routing problem with no skipping. 
    Returns metrics and the route for each vehicle.
    """
    manager = pywrapcp.RoutingIndexManager(len(data["distance_matrix"]),
                                           data["num_vehicles"],
                                           data["depot"])
    routing = pywrapcp.RoutingModel(manager)
    
    def distance_callback(from_index, to_index):
        return data["distance_matrix"][manager.IndexToNode(from_index)][manager.IndexToNode(to_index)]
    dist_cb_idx = routing.RegisterTransitCallback(distance_callback)
    routing.SetArcCostEvaluatorOfAllVehicles(dist_cb_idx)
    
    def demand_callback(from_index):
        return data["demands"][manager.IndexToNode(from_index)]
    demand_cb_idx = routing.RegisterUnaryTransitCallback(demand_callback)
    routing.AddDimensionWithVehicleCapacity(demand_cb_idx, 0,
                                            data["vehicle_capacities"],
                                            True, "Capacity")
    
    def time_callback(from_index, to_index):
        return data["time_matrix"][manager.IndexToNode(from_index)][manager.IndexToNode(to_index)]
    time_cb_idx = routing.RegisterTransitCallback(time_callback)
    routing.AddDimension(time_cb_idx, 30000, 24*60*60*1000, False, "Time")
    time_dim = routing.GetDimensionOrDie("Time")
    
    for loc_idx, tw in enumerate(data["time_windows"]):
        idx = manager.NodeToIndex(loc_idx)
        time_dim.CumulVar(idx).SetRange(tw[0]*60000, tw[1]*60000)
    
    # Pickup and Delivery
    for (pu_idx, do_idx) in data["pickups_deliveries"]:
        pu = manager.NodeToIndex(pu_idx)
        do = manager.NodeToIndex(do_idx)
        routing.AddPickupAndDelivery(pu, do)
        routing.solver().Add(routing.VehicleVar(pu) == routing.VehicleVar(do))
        routing.solver().Add(time_dim.CumulVar(pu) <= time_dim.CumulVar(do))
    
    # Accessibility constraints
    for pu_idx, (needs_bike, needs_wheelchair) in accessibility_requirements.items():
        allowed_vehicles = []
        for v, vehicle in enumerate(data["vehicles"]):
            if needs_bike and not vehicle["bike_capable"]:
                continue
            if needs_wheelchair and not vehicle["wheelchair_accessible"]:
                continue
            allowed_vehicles.append(v)
        if allowed_vehicles:
            routing.SetAllowedVehiclesForIndex(allowed_vehicles, manager.NodeToIndex(pu_idx))
    
    search_params = pywrapcp.DefaultRoutingSearchParameters()
    search_params.first_solution_strategy = routing_enums_pb2.FirstSolutionStrategy.PATH_CHEAPEST_ARC
    search_params.local_search_metaheuristic = routing_enums_pb2.LocalSearchMetaheuristic.GREEDY_DESCENT
    search_params.time_limit.FromSeconds(time_limit)
    search_params.log_search = True
    
    solution = routing.SolveWithParameters(search_params)
    if not solution:
        print("No solution found!")
        return None, None
    
    # Extract routes
    routes = []
    for v in range(data["num_vehicles"]):
        index = routing.Start(v)
        route_nodes = []
        while not routing.IsEnd(index):
            route_nodes.append(manager.IndexToNode(index))
            index = solution.Value(routing.NextVar(index))
        routes.append(route_nodes)
    
    # Calculate total distance
    total_dist_miles = solution.ObjectiveValue() / 1000.0
    sim_time_min = (total_dist_miles / 30.0) * 60.0
    
    # Gather vehicle details
    vehicle_details = []
    route_segment_distances = []
    for v, route in enumerate(routes):
        route_dist = 0.0
        non_depot_stops = 0
        load_over_route = []
        current_load = 0
        for i in range(len(route)-1):
            from_node = route[i]
            to_node = route[i+1]
            dist_leg = data["distance_matrix"][from_node][to_node] / 1000.0
            route_dist += dist_leg
            route_segment_distances.append(dist_leg)
            current_load += data["demands"][from_node]
            load_over_route.append(current_load)
            if from_node != data["depot"]:
                non_depot_stops += 1
        if route:
            last_node = route[-1]
            current_load += data["demands"][last_node]
            load_over_route.append(current_load)
            if last_node != data["depot"]:
                non_depot_stops += 1
        vehicle_details.append({
            "vehicle_id": v,
            "route_distance": route_dist,
            "stops": non_depot_stops,
            "load_list": load_over_route
        })
    
    bike_requests = sum(1 for (b, w) in accessibility_requirements.values() if b)
    wheelchair_requests = sum(1 for (b, w) in accessibility_requirements.values() if w)
    bike_vehicles = sum(1 for v in data["vehicles"] if v["bike_capable"])
    wheelchair_vehicles = sum(1 for v in data["vehicles"] if v["wheelchair_accessible"])
    
    metrics = {
        "distance_miles": total_dist_miles,
        "sim_time_min": sim_time_min,
        "vehicle_details": vehicle_details,
        "segment_distances": route_segment_distances,
        "bike_requests": bike_requests,
        "wheelchair_requests": wheelchair_requests,
        "bike_vehicles": bike_vehicles,
        "wheelchair_vehicles": wheelchair_vehicles
    }
    return metrics, routes

###############################################################################
# 4) VISUALIZATION (MAP)
###############################################################################
def visualize_routes(routes, data, locations, passenger_node_to_name,
                     accessibility_requirements, vehicle_colors):
    """
    Creates a Folium map with always-visible labels (DivIcon).
    """
    from folium.features import DivIcon
    depot_lat, depot_lon = locations[0][1], locations[0][2]
    m = folium.Map(location=[depot_lat, depot_lon], zoom_start=10)

    # Attempt to create an OSM graph for route directions
    try:
        G = ox.graph_from_point((depot_lat, depot_lon), dist=30000, network_type='drive')
        G = G.to_undirected()
    except Exception as e:
        print("OSMnx error, using direct coords only:", e)
        G = None

    for v_idx, route in enumerate(routes):
        color = vehicle_colors[v_idx % len(vehicle_colors)]
        coords = [(locations[node_idx][1], locations[node_idx][2]) for node_idx in route]
        folium.PolyLine(coords, color=color, weight=3).add_to(m)
        
        # Label each stop with a DivIcon
        for stop_order, node_idx in enumerate(route, start=1):
            lat, lon = locations[node_idx][1], locations[node_idx][2]
            dem = data["demands"][node_idx]
            if dem == 1:
                point_type = "Pickup"
                request_name = passenger_node_to_name.get(node_idx, "?")
            elif dem == -1:
                point_type = "Dropoff"
                request_name = passenger_node_to_name.get(node_idx, "?")
            else:
                point_type = "Depot"
                request_name = "Depot"
            
            needs_bike, needs_wheelchair = accessibility_requirements.get(node_idx, (False, False))
            icon_str = ""
            if needs_bike:
                icon_str += " "
            if needs_wheelchair:
                icon_str += " "
            
            label_text = f"Bus #{v_idx}, Stop #{stop_order}, {point_type} {request_name}{icon_str}"
            offset_lat = 0.00005
            offset_lon = 0.00005
            folium.map.Marker(
                location=(lat + offset_lat, lon + offset_lon),
                icon=DivIcon(
                    icon_size=(250, 36),
                    icon_anchor=(0, 0),
                    html=f"""
                    <div style="font-size: 10pt; 
                                color: white; 
                                background-color: {color}; 
                                opacity: 0.8;
                                border: 1px solid black; 
                                border-radius: 4px;
                                padding: 2px;">
                        {label_text}
                    </div>
                    """
                )
            ).add_to(m)
    
    return m

###############################################################################
# 5) BUILD RESULTS PAGE (HTML)
###############################################################################
def build_results_page(df, map_files, chart_file):
    html = "<html><head><title>Routing Results</title></head><body>"
    html += "<h1>Routing Test Results</h1>"
    html += "<p>Below are the aggregated metrics from our three case studies, each representing a distinct scenario in terms of number of requests, vehicles, and region size.</p>"
    html += "<h2>Aggregated Metrics</h2>"
    html += df.to_html(index=False)
    html += f"<h2>Comparison Chart (Basic)</h2><p>This bar chart compares total distance (left) and simulation time (right) across the scenarios. We see how the system scales with the number of requests and region size.</p>"
    html += f"<img src='{chart_file}' alt='Grouped Bar Chart' style='max-width:800px;'>"
    html += "<h2>Route Maps</h2><ul>"
    for cfg, mapfile in map_files.items():
        html += f"<li><a href='{mapfile}' target='_blank'>Config: {cfg}</a> - View the Folium map for this scenario.</li>"
    html += "</ul></body></html>"
    with open("results.html", "w") as f:
        f.write(html)
    print("Results page saved to results.html")

###############################################################################
# 6) ANALYZE AND PLOT (CHARTS WITH DESCRIPTIONS)
###############################################################################
def analyze_and_plot(all_runs_data):
    """
    Creates charts to illustrate:
      - Distance per vehicle
      - Stops per vehicle
      - Accessibility demand vs. supply
      - Distribution of route segment distances
      - Load over route order
    Each chart is accompanied by a textual explanation.
    """
    import numpy as np
    
    scenario_labels = []
    all_vehicle_distances = []
    all_vehicle_stops = []
    bike_requests_list = []
    wheelchair_requests_list = []
    bike_vehicles_list = []
    wheelchair_vehicles_list = []
    all_segment_dists = []
    load_charts_data = None
    load_charts_label = None
    
    # Gather data from each scenario
    for scenario_data in all_runs_data:
        scenario_label = scenario_data["Config"]
        scenario_labels.append(scenario_label)
        
        v_dists = [v["route_distance"] for v in scenario_data["vehicle_details"]]
        v_stops = [v["stops"] for v in scenario_data["vehicle_details"]]
        
        all_vehicle_distances.append(v_dists)
        all_vehicle_stops.append(v_stops)
        
        bike_requests_list.append(scenario_data["bike_requests"])
        wheelchair_requests_list.append(scenario_data["wheelchair_requests"])
        bike_vehicles_list.append(scenario_data["bike_vehicles"])
        wheelchair_vehicles_list.append(scenario_data["wheelchair_vehicles"])
        
        all_segment_dists.extend(scenario_data["segment_distances"])
        
        # We'll store the last scenario's data for the "Load Over Route" chart
        load_charts_data = scenario_data["vehicle_details"]
        load_charts_label = scenario_label
    
    # 1) Distance per Vehicle by Scenario
    fig1, axes1 = plt.subplots(len(scenario_labels), 1, figsize=(8, 4 * len(scenario_labels)))
    if len(scenario_labels) == 1:
        axes1 = [axes1]
    for idx, label in enumerate(scenario_labels):
        ax = axes1[idx]
        dist_list = all_vehicle_distances[idx]
        ax.bar(range(len(dist_list)), dist_list, color='skyblue')
        ax.set_title(f"Distance per Vehicle - {label}")
        ax.set_xlabel("Vehicle ID")
        ax.set_ylabel("Distance (miles)")
        # Explanation
        ax.text(0.5, 0.9, 
                "This chart shows how the total route distance is distributed among the vehicles.\n"
                "A balanced distribution suggests efficient load-sharing, while a large disparity\n"
                "may indicate certain vehicles are handling more requests.",
                transform=ax.transAxes, fontsize=8, verticalalignment='top', horizontalalignment='center')
    plt.tight_layout()
    plt.savefig("distance_per_vehicle.png")
    plt.show()
    
    # 2) Stops per Vehicle by Scenario
    fig2, axes2 = plt.subplots(len(scenario_labels), 1, figsize=(8, 4 * len(scenario_labels)))
    if len(scenario_labels) == 1:
        axes2 = [axes2]
    for idx, label in enumerate(scenario_labels):
        ax = axes2[idx]
        stops_list = all_vehicle_stops[idx]
        ax.bar(range(len(stops_list)), stops_list, color='orange')
        ax.set_title(f"Stops per Vehicle - {label}")
        ax.set_xlabel("Vehicle ID")
        ax.set_ylabel("Number of stops")
        # Explanation
        ax.text(0.5, 0.9,
                "This chart counts how many pickups/dropoffs each vehicle performed.\n"
                "It highlights whether requests are spread evenly or if certain vehicles\n"
                "handle most of the workload.",
                transform=ax.transAxes, fontsize=8, verticalalignment='top', horizontalalignment='center')
    plt.tight_layout()
    plt.savefig("stops_per_vehicle.png")
    plt.show()
    
    # 3) Accessibility Demand vs. Supply (Grouped Bar)
    x = range(len(scenario_labels))
    width = 0.2
    fig3, ax3 = plt.subplots(figsize=(10, 6))
    ax3.bar([xi - width for xi in x], bike_requests_list, width=width, label="Bike Requests", color='green')
    ax3.bar(x, wheelchair_requests_list, width=width, label="Wheelchair Requests", color='red')
    ax3.bar([xi + width for xi in x], bike_vehicles_list, width=width, label="Bike Vehicles", color='blue')
    ax3.bar([xi + 2*width for xi in x], wheelchair_vehicles_list, width=width, label="Wheelchair Vehicles", color='purple')
    ax3.set_xticks(list(x))
    ax3.set_xticklabels(scenario_labels, rotation=45)
    ax3.set_ylabel("Count")
    ax3.set_title("Accessibility Demand vs. Supply")
    ax3.legend()
    # Explanation
    ax3.text(0.5, 1.03,
             "This grouped bar chart compares the total number of accessibility requests\n"
             "(bike & wheelchair) to the number of vehicles capable of meeting those needs.\n"
             "It shows if the system can handle the accessibility demand in each scenario.",
             transform=ax3.transAxes, fontsize=9, verticalalignment='bottom', horizontalalignment='center')
    plt.tight_layout()
    plt.savefig("accessibility_demand_vs_supply.png")
    plt.show()
    
    # 4) Distribution of Route Segment Distances (Histogram)
    fig4, ax4 = plt.subplots(figsize=(8, 6))
    ax4.hist(all_segment_dists, bins=10, color='gray', edgecolor='black')
    ax4.set_title("Distribution of Route Segment Distances")
    ax4.set_xlabel("Segment Distance (miles)")
    ax4.set_ylabel("Frequency")
    # Explanation
    ax4.text(0.5, 0.9,
             "This histogram shows how long individual route segments are across all scenarios.\n"
             "Many short segments indicate closer pickups/dropoffs, while a tail of longer segments\n"
             "suggests some rides are significantly farther apart.",
             transform=ax4.transAxes, fontsize=8, verticalalignment='top', horizontalalignment='center')
    plt.tight_layout()
    plt.savefig("segment_distance_hist.png")
    plt.show()
    
    # 5) Load Over Route Order (Line Chart) - from the last scenario
    if load_charts_data is not None:
        fig5, ax5 = plt.subplots(figsize=(10, 6))
        for vinfo in load_charts_data:
            v_id = vinfo["vehicle_id"]
            load_list = vinfo["load_list"]
            ax5.plot(range(len(load_list)), load_list, marker='o', label=f"Vehicle {v_id}")
        ax5.set_title(f"Load Over Route Order - {load_charts_label}")
        ax5.set_xlabel("Stop Index in Route")
        ax5.set_ylabel("Passengers Onboard")
        ax5.legend()
        # Explanation
        ax5.text(0.5, 1.03,
                 "This line chart displays how the passenger load changes stop by stop.\n"
                 "It helps identify if a vehicle quickly fills up or remains under capacity.",
                 transform=ax5.transAxes, fontsize=9, verticalalignment='bottom', horizontalalignment='center')
        plt.tight_layout()
        plt.savefig("load_over_route.png")
        plt.show()

###############################################################################
# 7) MAIN: RUN CASE STUDIES, PLOT, & GENERATE RESULTS
###############################################################################
def main():
    all_runs = []
    map_files = {}
    
    for config in TEST_CONFIGS:
        num_requests = config["num_requests"]
        num_vehicles = config["num_vehicles"]
        region_size = config["region_size"]
        config_label = f"Req{num_requests}_Veh{num_vehicles}_Reg{region_size}"
        print(f"\n=== Running test config: {config_label} ===")
        
        # Create random data for this scenario
        data, locations, passenger_node_to_name, accessibility_requirements, requests = \
            create_rural_data(num_requests=num_requests,
                              num_vehicles=num_vehicles,
                              region_size=region_size)
        
        # Solve routing
        metrics, routes = solve_routing(data, locations, passenger_node_to_name,
                                        accessibility_requirements,
                                        fs_strategy="PCA", ls_heuristic="GLS",
                                        time_limit=TIME_LIMIT_SEC)
        if metrics is None:
            continue
        
        # Store scenario-level info
        metrics.update({
            "Config": config_label,
            "NumRequests": num_requests,
            "NumVehicles": num_vehicles,
            "RegionSize": region_size,
            "requests": requests
        })
        all_runs.append(metrics)
        
        # Create a map for each scenario
        ts = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")
        mapfile = f"routes_{config_label}_{ts}.html"
        vehicle_colors = ['red', 'blue', 'green', 'purple', 'orange']
        my_map = visualize_routes(routes, data, locations, passenger_node_to_name,
                                  accessibility_requirements, vehicle_colors)
        my_map.save(mapfile)
        map_files[config_label] = mapfile
        
        # Display inline if in Jupyter
        if IN_JUPYTER:
            display(my_map)
    
    if not all_runs:
        print("No solutions found for any configuration!")
        return
    
    # Build a DataFrame of the main metrics for each scenario
    df = pd.DataFrame([
        {
            "Config": r["Config"],
            "NumRequests": r["NumRequests"],
            "NumVehicles": r["NumVehicles"],
            "RegionSize": r["RegionSize"],
            "distance_miles": r["distance_miles"],
            "sim_time_min": r["sim_time_min"]
        } for r in all_runs
    ])
    
    print("\n--- Raw Results (All Configurations) ---")
    print(df)
    
    # Basic bar charts for total distance & sim time
    df_indexed = df.set_index("Config")
    fig, axes = plt.subplots(1, 2, figsize=(14, 6))
    df_indexed["distance_miles"].plot(kind="bar", ax=axes[0], rot=45, width=0.5)
    axes[0].set_title("Total Distance (miles)")
    axes[0].set_ylabel("Miles")
    
    df_indexed["sim_time_min"].plot(kind="bar", ax=axes[1], rot=45, width=0.5)
    axes[1].set_title("Simulation Time (min)")
    axes[1].set_ylabel("Minutes")
    plt.tight_layout()
    chart_filename = "grouped_comparison.png"
    plt.savefig(chart_filename)
    plt.show()
    print(f"Basic grouped bar chart saved to {chart_filename}")
    
    # Build HTML results page
    build_results_page(df, map_files, chart_filename)
    
    # Create additional analysis & charts with textual explanations
    analyze_and_plot(all_runs)

if __name__ == "__main__":
    main()
```

```python
import sys
import random
import datetime
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np # Make sure numpy is imported
import os

# Use perf_counter for high-resolution timing
from time import perf_counter

from ortools.constraint_solver import routing_enums_pb2, pywrapcp
from haversine import haversine, Unit
import folium
# import osmnx as ox # Keep commented if not essential and causing issues
# import networkx as nx # Keep commented if not essential

# For Jupyter display if needed
IN_JUPYTER = 'ipykernel' in sys.modules
if IN_JUPYTER:
    from IPython.display import display

###############################################################################
# 1) SCENARIO DEFINITIONS (Case Studies)
###############################################################################
TEST_CONFIGS = [
    # Case Study A: small region, fewer requests
    {"num_requests": 10, "num_vehicles": 2, "region_size": 0.05, "name": "Case A"},
    # Case Study B: moderate region, moderate requests
    {"num_requests": 20, "num_vehicles": 4, "region_size": 0.1, "name": "Case B"},
    # Case Study C: larger region, more requests
    {"num_requests": 30, "num_vehicles": 6, "region_size": 0.2, "name": "Case C"},
]
TIME_LIMIT_SEC = 30

# Define output directory
OUTPUT_DIR = "rural_routing_output_structured"
if not os.path.exists(OUTPUT_DIR):
    os.makedirs(OUTPUT_DIR)
print(f"Output will be saved in: {OUTPUT_DIR}")

###############################################################################
# 2) DATA CREATION (RANDOM) - Unchanged from your working version
###############################################################################
def create_rural_data(num_requests=8, num_vehicles=3, region_size=0.1):
    """
    Generates a dial-a-ride problem with wide time windows and large capacities.
    Randomness ensures each run is unique. We do not fix seeds here.
    """
    # --- Using the exact function you provided that worked ---
    center_lat, center_lon = 30.1897, -82.6393
    lat_range = region_size # Using original interpretation for now
    lon_range = region_size # Using original interpretation for now

    locations = [("Depot", center_lat, center_lon)]
    requests = []
    accessibility_requirements = {} # Maps pickup node index to (needs_bike, needs_wheelchair)

    for i in range(num_requests):
        pickup_lat = center_lat + random.uniform(-lat_range, lat_range)
        pickup_lon = center_lon + random.uniform(-lon_range, lon_range)
        dropoff_lat = center_lat + random.uniform(-lat_range, lat_range)
        dropoff_lon = center_lon + random.uniform(-lon_range, lon_range)
        # Optional: Add minimum distance check if desired
        # while haversine((pickup_lat, pickup_lon), (dropoff_lat, dropoff_lon)) < 0.5:
        #      dropoff_lat = center_lat + random.uniform(-lat_range, lat_range)
        #      dropoff_lon = center_lon + random.uniform(-lon_range, lon_range)

        # Full-day time windows (0 to 1440 minutes)
        window = (0, 1440)

        # Random accessibility needs (50% chance for each)
        needs_bike = (random.random() < 0.5)
        needs_wheelchair = (random.random() < 0.5)

        req = {
            "name": f"R{i+1}",
            "pickup_lat": pickup_lat,
            "pickup_lon": pickup_lon,
            "dropoff_lat": dropoff_lat,
            "dropoff_lon": dropoff_lon,
            "time_window": window,
            "needs_bike": needs_bike,
            "needs_wheelchair": needs_wheelchair
        }
        requests.append(req)

    # Build location list for OR-Tools (Depot, P1, D1, P2, D2, ...)
    node_index = 1
    for i, req in enumerate(requests): # Use enumerate to get index i
        # pickup
        locations.append((f"{req['name']}_pickup", req["pickup_lat"], req["pickup_lon"]))
        pickup_node_idx = node_index
        node_index += 1
        # dropoff
        locations.append((f"{req['name']}_dropoff", req["dropoff_lat"], req["dropoff_lon"]))
        node_index += 1
        # Store accessibility for pickup node index (matches original logic: 1+2*i)
        accessibility_requirements[1 + 2*i] = (req["needs_bike"], req["needs_wheelchair"])

    data = {}
    data["num_vehicles"] = num_vehicles
    data["depot"] = 0
    # Large capacity
    data["vehicle_capacities"] = [9999] * num_vehicles

    num_locs = len(locations)
    dist_mat = np.zeros((num_locs, num_locs))
    for i in range(num_locs):
        for j in range(num_locs):
            coord1 = (locations[i][1], locations[i][2])
            coord2 = (locations[j][1], locations[j][2])
            dist_mat[i][j] = haversine(coord1, coord2, unit=Unit.MILES)

    # Scale distance to integers (meters or similar)
    data["distance_matrix"] = (dist_mat * 1000).astype(int).tolist()

    # Time matrix: assume average speed 30 mph
    avg_speed_mph = 30.0
    time_mat_minutes = (dist_mat / avg_speed_mph) * 60.0
    # Scale time to integers (e.g., seconds * 10 or milliseconds)
    data["time_matrix"] = (time_mat_minutes * 1000).astype(int).tolist()

    # Demands: +1 for pickup, -1 for dropoff
    data["demands"] = [0] * num_locs
    data["pickups_deliveries"] = []
    data["time_windows"] = [] # List of (start_time_min, end_time_min) * 1000

    # Depot time window (full day in scaled units) -> Use original scale
    # Original code used minutes for TW list, but scaled CumulVar range later. Let's stick to minutes here.
    data["time_windows"].append((0, 1440))

    passenger_node_to_name = {0: "Depot"} # Map node index back to request name
    for i, req in enumerate(requests):
        pu_idx = 1 + 2*i
        do_idx = pu_idx + 1
        data["pickups_deliveries"].append([pu_idx, do_idx])
        data["demands"][pu_idx] = 1
        data["demands"][do_idx] = -1
        # Use unique names for pickup/dropoff in mapping for clarity
        passenger_node_to_name[pu_idx] = req["name"] # Store only request name
        passenger_node_to_name[do_idx] = req["name"] # Store only request name

        # Add time windows (original scale: minutes)
        data["time_windows"].append(req["time_window"]) # Pickup TW
        data["time_windows"].append(req["time_window"]) # Dropoff TW

    # Vehicle accessibility: random assignment
    data["vehicles"] = []
    for v in range(num_vehicles):
        bike_capable = (random.random() < 0.5)
        wheelchair_accessible = (random.random() < 0.5)
        data["vehicles"].append({
            "id": v, # Add ID
            "bike_capable": bike_capable,
            "wheelchair_accessible": wheelchair_accessible
        })

    return data, locations, passenger_node_to_name, accessibility_requirements, requests

###############################################################################
# 3) SOLVE ROUTING
###############################################################################
def solve_routing(data, locations, passenger_node_to_name, accessibility_requirements,
                  fs_strategy_str="PATH_CHEAPEST_ARC", ls_heuristic_str="GREEDY_DESCENT", time_limit=60):
    """
    Solves the routing problem, allowing requests to be dropped for a penalty
    if they are infeasible (e.g., due to accessibility).
    Uses ActiveVar to correctly check if nodes were dropped.
    Returns metrics (including dropped count), routes, and solve duration.
    """
    manager = pywrapcp.RoutingIndexManager(len(data["distance_matrix"]),
                                           data["num_vehicles"],
                                           data["depot"])
    routing = pywrapcp.RoutingModel(manager)

    # --- Penalty for Dropping Requests ---
    penalty = 1_000_000_000

    # --- Distance Dimension ---
    def distance_callback(from_index, to_index):
        from_node = manager.IndexToNode(from_index)
        to_node = manager.IndexToNode(to_index)
        return data["distance_matrix"][from_node][to_node]
    dist_cb_idx = routing.RegisterTransitCallback(distance_callback)
    routing.SetArcCostEvaluatorOfAllVehicles(dist_cb_idx)

    # --- Capacity Dimension ---
    def demand_callback(from_index):
        from_node = manager.IndexToNode(from_index)
        return data["demands"][from_node]
    demand_cb_idx = routing.RegisterUnaryTransitCallback(demand_callback)
    routing.AddDimensionWithVehicleCapacity(
        demand_cb_idx, 0, data["vehicle_capacities"], True, "Capacity")

    # --- Time Dimension ---
    def time_callback(from_index, to_index):
        from_node = manager.IndexToNode(from_index)
        to_node = manager.IndexToNode(to_index)
        return data["time_matrix"][from_node][to_node]
    time_cb_idx = routing.RegisterTransitCallback(time_callback)
    horizon = 24 * 60 * 60 * 1000
    routing.AddDimension(time_cb_idx, horizon, horizon, False, "Time")
    time_dim = routing.GetDimensionOrDie("Time")

    # Add time window constraints (scaled)
    for loc_idx, tw in enumerate(data["time_windows"]):
        if loc_idx == data["depot"]: continue
        index = manager.NodeToIndex(loc_idx)
        tw_start_scaled = tw[0] * 60 * 1000
        tw_end_scaled = tw[1] * 60 * 1000
        time_dim.CumulVar(index).SetRange(tw_start_scaled, tw_end_scaled)

    # Depot time windows
    depot_tw = data["time_windows"][data["depot"]]
    depot_start_scaled = depot_tw[0] * 60 * 1000
    depot_end_scaled = depot_tw[1] * 60 * 1000
    for v in range(data["num_vehicles"]):
        start_index = routing.Start(v)
        end_index = routing.End(v)
        time_dim.CumulVar(start_index).SetRange(depot_start_scaled, depot_end_scaled)
        time_dim.CumulVar(end_index).SetRange(depot_start_scaled, depot_end_scaled)

    # --- Pickup and Delivery Constraints with Penalties ---
    request_indices = [] # Stores pairs of routing indices (pickup_index, delivery_index)
    pickup_node_ids_original = [] # Stores original node ids (1, 3, 5, ...) for pickups

    for i, (pu_idx, do_idx) in enumerate(data["pickups_deliveries"]):
        pickup_node_index = manager.NodeToIndex(pu_idx)
        delivery_node_index = manager.NodeToIndex(do_idx)
        request_indices.append((pickup_node_index, delivery_node_index))
        pickup_node_ids_original.append(pu_idx) # Store the original ID

        # --- Add Disjunction for Pickup ---
        # Correct call without name parameter
        routing.AddDisjunction([pickup_node_index], penalty)
        # No need to store disjunction_index if using ActiveVar check later

        # --- Add standard P&D constraints ---
        routing.AddPickupAndDelivery(pickup_node_index, delivery_node_index)
        routing.solver().Add(routing.VehicleVar(pickup_node_index) == routing.VehicleVar(delivery_node_index))
        routing.solver().Add(time_dim.CumulVar(pickup_node_index) <= time_dim.CumulVar(delivery_node_index))

    # --- Accessibility Constraints ---
    for pu_idx, (needs_bike, needs_wheelchair) in accessibility_requirements.items():
        pickup_node_index_acc = manager.NodeToIndex(pu_idx)
        allowed_vehicles = []
        for v_idx, vehicle in enumerate(data["vehicles"]):
            can_serve = True
            if needs_bike and not vehicle["bike_capable"]: can_serve = False
            if needs_wheelchair and not vehicle["wheelchair_accessible"]: can_serve = False
            if can_serve: allowed_vehicles.append(v_idx)

        if not allowed_vehicles:
             print(f"INFO: Request at node {pu_idx} ({passenger_node_to_name.get(pu_idx)}) has accessibility needs no vehicle can meet. Will be dropped.")

        routing.SetAllowedVehiclesForIndex(allowed_vehicles, pickup_node_index_acc)
        do_idx_acc = pu_idx + 1
        if do_idx_acc < len(data["distance_matrix"]):
             delivery_node_index_acc = manager.NodeToIndex(do_idx_acc)
             routing.SetAllowedVehiclesForIndex(allowed_vehicles, delivery_node_index_acc)

    # --- Solver Settings ---
    search_params = pywrapcp.DefaultRoutingSearchParameters()
    fs_strategy_enum = getattr(routing_enums_pb2.FirstSolutionStrategy, fs_strategy_str, routing_enums_pb2.FirstSolutionStrategy.AUTOMATIC)
    search_params.first_solution_strategy = fs_strategy_enum
    ls_heuristic_enum = getattr(routing_enums_pb2.LocalSearchMetaheuristic, ls_heuristic_str, routing_enums_pb2.LocalSearchMetaheuristic.AUTOMATIC)
    search_params.local_search_metaheuristic = ls_heuristic_enum
    search_params.time_limit.FromSeconds(time_limit)
    search_params.log_search = False

    # --- Solve ---
    print(f"  Solving with FS={fs_strategy_str}, LS={ls_heuristic_str}, Limit={time_limit}s (with penalties)...")
    start_time = perf_counter()
    solution = routing.SolveWithParameters(search_params)
    solve_duration = perf_counter() - start_time
    print(f"  Solve finished in {solve_duration:.2f} seconds.")

    # --- Extract Routes and Metrics ---
    if not solution:
        print("  No solution found (even with penalties)!")
        return None, None, solve_duration

    # *** CORRECTED Check for dropped requests using ActiveVar ***
    dropped_requests_count = 0
    dropped_request_names = []
    served_pickup_indices = set()

    # Loop through the original pickup node IDs we stored
    for pu_node_idx in pickup_node_ids_original:
         pickup_index = manager.NodeToIndex(pu_node_idx) # Get the routing index

         # Check the value of the ActiveVar for the pickup node in the solution.
         # If ActiveVar(node).Value() == 0, the node is inactive (dropped).
         if solution.Value(routing.ActiveVar(pickup_index)) == 0:
             # Node is inactive -> request was dropped
             dropped_requests_count += 1
             dropped_request_names.append(passenger_node_to_name.get(pu_node_idx, f"Node_{pu_node_idx}"))
         else:
             # Node is active -> request was served
             served_pickup_indices.add(pu_node_idx)


    print(f"  Number of dropped requests: {dropped_requests_count}")
    if dropped_request_names:
        print(f"  Dropped request names: {', '.join(dropped_request_names)}")

    # Extract actual routes and calculate metrics excluding penalties
    routes = []
    total_dist_scaled = 0
    max_route_time_scaled = 0
    vehicle_details = []
    route_segment_distances = []

    for v in range(data["num_vehicles"]):
        index = routing.Start(v)
        route_nodes = []
        route_dist_vehicle_scaled = 0
        current_load = 0
        load_over_route = [0]
        stop_times = {}

        while not routing.IsEnd(index):
            node_idx = manager.IndexToNode(index)
            route_nodes.append(node_idx)
            stop_times[node_idx] = time_dim.CumulVar(index).Value()

            current_load += data["demands"][node_idx]
            load_over_route.append(current_load)

            previous_index = index
            index = solution.Value(routing.NextVar(index))

            arc_cost = routing.GetArcCostForVehicle(previous_index, index, v)
            route_dist_vehicle_scaled += arc_cost

            from_node_seg = manager.IndexToNode(previous_index)
            to_node_seg = manager.IndexToNode(index)
            # Basic check to avoid potential index error if to_node_seg is invalid (shouldn't happen)
            if 0 <= from_node_seg < len(data["distance_matrix"]) and 0 <= to_node_seg < len(data["distance_matrix"]):
                seg_dist_miles = data["distance_matrix"][from_node_seg][to_node_seg] / 1000.0
                route_segment_distances.append(seg_dist_miles)
            else:
                print(f"WARN: Invalid node index encountered during segment distance calculation: {from_node_seg} -> {to_node_seg}")


        final_node_idx = manager.IndexToNode(index)
        route_nodes.append(final_node_idx)
        stop_times[final_node_idx] = time_dim.CumulVar(index).Value()

        routes.append(route_nodes)
        total_dist_scaled += route_dist_vehicle_scaled

        non_depot_stops = sum(1 for node in route_nodes if node != data["depot"])
        route_dist_miles = route_dist_vehicle_scaled / 1000.0
        vehicle_details.append({
            "vehicle_id": v,
            "route_distance": route_dist_miles,
            "stops": non_depot_stops,
            "load_list": load_over_route,
            "route_nodes": route_nodes,
            "stop_times_ms": stop_times
        })

        route_end_time_scaled = time_dim.CumulVar(index).Value()
        if route_end_time_scaled > max_route_time_scaled:
            max_route_time_scaled = route_end_time_scaled

    # Final metric calculations
    total_dist_miles = total_dist_scaled / 1000.0
    sim_time_min = max_route_time_scaled / (60 * 1000.0)

    # Gather accessibility stats
    bike_requests = sum(1 for (b, w) in accessibility_requirements.values() if b)
    wheelchair_requests = sum(1 for (b, w) in accessibility_requirements.values() if w)
    bike_vehicles = sum(1 for v in data["vehicles"] if v["bike_capable"])
    wheelchair_vehicles = sum(1 for v in data["vehicles"] if v["wheelchair_accessible"])

    # Prepare final metrics dictionary
    metrics = {
        "distance_miles": total_dist_miles,
        "sim_time_min": sim_time_min,
        "solve_duration_sec": solve_duration,
        "vehicle_details": vehicle_details,
        "segment_distances": route_segment_distances,
        "bike_requests": bike_requests,
        "wheelchair_requests": wheelchair_requests,
        "bike_vehicles": bike_vehicles,
        "wheelchair_vehicles": wheelchair_vehicles,
        "dropped_requests_count": dropped_requests_count,
        "dropped_request_names": dropped_request_names,
        "status": "Success"
    }
    return metrics, routes, solve_duration


###############################################################################
# 4) VISUALIZATION (MAP) - MODIFIED for Clarity
###############################################################################
def visualize_routes_clearer(routes, data, locations, passenger_node_to_name,
                             accessibility_requirements, vehicle_colors, map_center):
    """
    Creates a Folium map using standard markers with popups and tooltips
    instead of DivIcons for less clutter.
    """
    if not folium:
        print("Folium not installed, cannot create map.")
        return None

    m = folium.Map(location=map_center, zoom_start=12) # Use provided center

    # Add Depot marker (distinct style)
    depot_lat, depot_lon = locations[data["depot"]][1], locations[data["depot"]][2]
    folium.Marker(
        location=[depot_lat, depot_lon],
        popup="<b>Depot</b>",
        tooltip="Depot",
        icon=folium.Icon(color='black', icon='home', prefix='fa') # FontAwesome home icon
    ).add_to(m)

    # Define icons for pickup and dropoff
    pickup_icon = folium.Icon(color='green', icon='arrow-up', prefix='fa')
    dropoff_icon = folium.Icon(color='red', icon='arrow-down', prefix='fa')

    for v_idx, route in enumerate(routes):
        color = vehicle_colors[v_idx % len(vehicle_colors)]
        route_coords = []
        if not route or len(route) <= 1: # Skip empty routes
             continue

        # Collect coordinates for polyline
        for node_idx in route:
             route_coords.append((locations[node_idx][1], locations[node_idx][2]))

        # Draw lines connecting stops in order for this vehicle
        folium.PolyLine(route_coords, color=color, weight=2.5, opacity=0.7).add_to(m)

        # Add Markers for each stop (excluding depot start/end in markers loop)
        for stop_order, node_idx in enumerate(route):
            if node_idx == data["depot"]: continue # Skip depot, already marked

            lat, lon = locations[node_idx][1], locations[node_idx][2]
            loc_name = locations[node_idx][0] # e.g., R1_pickup
            request_name = passenger_node_to_name.get(node_idx, "?") # e.g., R1

            # Determine type and icon
            is_pickup = data["demands"][node_idx] > 0
            point_type = "Pickup" if is_pickup else "Dropoff"
            marker_icon = pickup_icon if is_pickup else dropoff_icon
            marker_color = 'green' if is_pickup else 'red' # Color for standard circle marker if icon fails

            # Check accessibility (relevant for pickup node)
            needs_bike, needs_wheelchair = accessibility_requirements.get(node_idx, (False, False))
            acc_str = ""
            # Show accessibility requirements only at pickup for clarity
            if is_pickup:
                if needs_bike: acc_str += " Needs Bike"
                if needs_wheelchair: acc_str += " Needs Wheelchair"
            acc_html = f"<br><i>{acc_str.strip()}</i>" if acc_str else ""

            # Create Tooltip (simple hover text)
            tooltip_text = f"V{v_idx} Stop {stop_order}: {point_type} {request_name}"

            # Create Popup (detailed click text)
            popup_html = f"""
            <b>Vehicle:</b> {v_idx}<br>
            <b>Stop:</b> {stop_order}<br>
            <b>Type:</b> {point_type}<br>
            <b>Request:</b> {request_name}
            {acc_html}
            """

            # Add the marker
            folium.Marker(
                location=(lat, lon),
                popup=folium.Popup(popup_html, max_width=200),
                tooltip=tooltip_text,
                icon=folium.Icon(color=marker_color, icon='info-sign') # Simple icon, color indicates type
                # Alternative: Use the fa icons defined earlier
                # icon=pickup_icon if is_pickup else dropoff_icon
            ).add_to(m)

    return m


###############################################################################
# 5) BUILD RESULTS PAGE (HTML) - MODIFIED for Structure
###############################################################################
def build_structured_results_page(all_scenario_results, summary_df, chart_files):
    """ Builds an HTML page structured by case study. """

    html = f"""
    <html>
    <head>
        <title>Rural Transit Routing Results</title>
        <style>
            body {{ font-family: sans-serif; margin: 20px; }}
            h1 {{ color: #333; }}
            h2 {{ color: #555; border-bottom: 1px solid #ccc; padding-bottom: 5px; margin-top: 30px; }}
            h3 {{ color: #666; margin-top: 20px; }}
            table {{ border-collapse: collapse; margin-top: 10px; }}
            th, td {{ border: 1px solid #ddd; padding: 8px; text-align: left; }}
            th {{ background-color: #f2f2f2; }}
            img {{ max-width: 700px; height: auto; border: 1px solid #ccc; margin-top: 10px; margin-bottom: 10px; }}
            .metrics-table td {{ min-width: 100px; }}
            .summary-section {{ background-color: #f9f9f9; padding: 15px; border: 1px solid #eee; margin-top: 40px; }}
        </style>
    </head>
    <body>
    <h1>Rural Transit Routing Simulation Report</h1>
    <p>This report details the results of routing simulations for different rural transit scenarios.</p>
    <p><i>Report generated on: {datetime.datetime.now().strftime('%Y-%m-%d %H:%M:%S')}</i></p>
    """

    # --- Loop through each scenario ---
    for config_label, results in all_scenario_results.items():
        # Find the matching config name (e.g., "Case A")
        config_name = "Unknown Scenario"
        for cfg in TEST_CONFIGS:
            if f"Req{cfg['num_requests']}_Veh{cfg['num_vehicles']}_Reg{cfg['region_size']:.2f}" == config_label:
                 config_name = cfg.get('name', config_label)
                 break

        html += f"<h2>{config_name}: {config_label}</h2>\n"

        if results['status'] == "Success":
            html += f"<p><b>Status:</b> Solved Successfully</p>\n"

            # Scenario Metrics Table
            html += "<h3>Key Metrics</h3>\n"
            html += "<table class='metrics-table'>\n"
            html += f"<tr><th>Metric</th><th>Value</th></tr>\n"
            html += f"<tr><td>Total Distance</td><td>{results['distance_miles']:.2f} miles</td></tr>\n"
            html += f"<tr><td>Max Route Duration</td><td>{results['sim_time_min']:.2f} min</td></tr>\n"
            html += f"<tr><td>Solve Time</td><td>{results['solve_duration_sec']:.2f} s</td></tr>\n"
            html += f"<tr><td>Bike Requests / Vehicles</td><td>{results['bike_requests']} / {results['bike_vehicles']}</td></tr>\n"
            html += f"<tr><td>Wheelchair Requests / Vehicles</td><td>{results['wheelchair_requests']} / {results['wheelchair_vehicles']}</td></tr>\n"
            html += "</table>\n"

            # Link to Map
            map_file = chart_files.get(config_label, {}).get('map')
            if map_file and os.path.exists(map_file):
                html += f"<h3>Route Map</h3>\n"
                html += f"<p><a href='{os.path.basename(map_file)}' target='_blank'>View Interactive Map for {config_name}</a></p>\n"
            else:
                 html += f"<p><i>Map file not found or not generated.</i></p>\n"

            # Embed Scenario-Specific Charts
            dist_chart = chart_files.get(config_label, {}).get('distance_per_vehicle')
            stops_chart = chart_files.get(config_label, {}).get('stops_per_vehicle')
            load_chart = chart_files.get(config_label, {}).get('load_over_route') # Assuming one load chart per scenario now

            if dist_chart and os.path.exists(dist_chart):
                html += f"<h3>Workload: Distance per Vehicle</h3>\n"
                html += f"<img src='{os.path.basename(dist_chart)}' alt='Distance per Vehicle for {config_name}'><br/>\n"

            if stops_chart and os.path.exists(stops_chart):
                html += f"<h3>Workload: Stops per Vehicle</h3>\n"
                html += f"<img src='{os.path.basename(stops_chart)}' alt='Stops per Vehicle for {config_name}'><br/>\n"

            if load_chart and os.path.exists(load_chart):
                html += f"<h3>Vehicle Load Profile</h3>\n"
                html += f"<img src='{os.path.basename(load_chart)}' alt='Load Profile for {config_name}'><br/>\n"

        else:
            html += f"<p><b>Status:</b> Failed to find solution.</p>\n"
            html += f"<p>Solve Time: {results['solve_duration_sec']:.2f} s</p>\n"
            # Display demand vs supply even on failure
            html += "<table class='metrics-table'>\n"
            html += f"<tr><th>Metric</th><th>Value</th></tr>\n"
            html += f"<tr><td>Bike Requests / Vehicles</td><td>{results['bike_requests']} / {results['bike_vehicles']}</td></tr>\n"
            html += f"<tr><td>Wheelchair Requests / Vehicles</td><td>{results['wheelchair_requests']} / {results['wheelchair_vehicles']}</td></tr>\n"
            html += "</table>\n"


    # --- Summary Section ---
    html += "<div class='summary-section'>\n"
    html += "<h2>Overall Summary & Comparison</h2>\n"

    # Summary Table
    html += "<h3>Aggregated Metrics Across Scenarios</h3>\n"
    html += summary_df.to_html(index=False, border=1, classes='metrics-table') + "\n"

    # Embed Aggregate Charts
    comp_chart = chart_files.get('aggregate', {}).get('comparison_summary')
    access_chart = chart_files.get('aggregate', {}).get('accessibility_demand_vs_supply')
    hist_chart = chart_files.get('aggregate', {}).get('segment_distance_hist')

    if comp_chart and os.path.exists(comp_chart):
        html += f"<h3>Performance Comparison Chart</h3>\n"
        html += f"<p>Compares total distance and max duration for successful scenarios.</p>\n"
        html += f"<img src='{os.path.basename(comp_chart)}' alt='Overall Comparison Chart'><br/>\n"

    if access_chart and os.path.exists(access_chart):
        html += f"<h3>Accessibility Demand vs. Supply</h3>\n"
        html += f"<p>Compares requests needing accessibility features vs. available equipped vehicles across all scenarios.</p>\n"
        html += f"<img src='{os.path.basename(access_chart)}' alt='Accessibility Demand vs Supply Chart'><br/>\n"

    if hist_chart and os.path.exists(hist_chart):
        html += f"<h3>Inter-Stop Distance Distribution</h3>\n"
        html += f"<p>Histogram showing the frequency of distances between consecutive stops, aggregated across all successful scenarios.</p>\n"
        html += f"<img src='{os.path.basename(hist_chart)}' alt='Segment Distance Histogram'><br/>\n"

    html += "<h3>Conclusions</h3>\n"
    html += "<p><i>[Add your analysis and conclusions here based on the results...]</i></p>\n"
    html += "<ul><li>Observe how total distance and duration scale with requests/region size.</li>"
    html += "<li>Analyze workload balance using the per-vehicle distance/stop charts for each case.</li>"
    html += "<li>Check the accessibility chart to see if vehicle supply meets demand.</li>"
    html += "<li>Examine the distance histogram for insights into route sparsity.</li>"
    html += "<li>Review vehicle load profiles to check capacity utilization.</li></ul>\n"

    html += "</div>\n" # End summary-section
    html += "</body></html>"

    results_filename = os.path.join(OUTPUT_DIR, "rural_routing_structured_report.html")
    try:
        with open(results_filename, "w") as f:
            f.write(html)
        print(f"\nStructured HTML report saved to {results_filename}")
    except Exception as e:
        print(f"Error writing structured HTML results page: {e}")


###############################################################################
# 6) ANALYZE AND PLOT - MODIFIED for Structure and Clarity
###############################################################################
def analyze_and_plot_structured(all_scenario_results):
    """
    Creates charts:
      - Scenario-specific: Distance/Stops per vehicle, Load profile
      - Aggregate: Accessibility D/S, Segment Distance Histogram
    Saves charts to files and returns a dictionary of filenames.
    Adds background boxes to text and adjusts layout.
    """
    chart_files = {'aggregate': {}, 'scenarios': {}} # Store filenames

    all_segment_dists = []
    successful_runs_data = {k: v for k, v in all_scenario_results.items() if v['status'] == 'Success'}
    if not successful_runs_data:
        print("No successful runs to generate detailed plots for.")
        # Still generate accessibility if possible? Need data from all runs.
        all_runs_data_list = list(all_scenario_results.values()) # Use all for accessibility
    else:
         all_runs_data_list = list(successful_runs_data.values()) # Use only successful for histogram, etc.

    # Data for aggregate plots
    scenario_labels_all = [r["Config"] for r in all_scenario_results.values()]
    bike_requests_list_all = [r["bike_requests"] for r in all_scenario_results.values()]
    wheelchair_requests_list_all = [r["wheelchair_requests"] for r in all_scenario_results.values()]
    bike_vehicles_list_all = [r["bike_vehicles"] for r in all_scenario_results.values()]
    wheelchair_vehicles_list_all = [r["wheelchair_vehicles"] for r in all_scenario_results.values()]

    # --- Generate Scenario-Specific Charts ---
    print("\n--- Generating Scenario-Specific Charts ---")
    for config_label, scenario_data in successful_runs_data.items():
        print(f"  Processing charts for {config_label}...")
        chart_files['scenarios'][config_label] = {}
        vehicle_details = scenario_data["vehicle_details"]
        num_vehicles = len(vehicle_details)
        if num_vehicles == 0:
            print(f"    Skipping {config_label}: No vehicle routes found.")
            continue

        # Chart 1a: Distance per Vehicle (for this scenario)
        fig1, ax1 = plt.subplots(figsize=(max(6, num_vehicles*1.5), 4)) # Adjust width
        dists = [v["route_distance"] for v in vehicle_details]
        v_ids = [f"V{v['vehicle_id']}" for v in vehicle_details]
        ax1.bar(v_ids, dists, color='skyblue')
        ax1.set_title(f"Distance per Vehicle - {config_label}")
        ax1.set_xlabel("Vehicle ID")
        ax1.set_ylabel("Distance (miles)")
        ax1.text(0.98, 0.95, "Workload distribution by distance.",
                 transform=ax1.transAxes, fontsize=9, verticalalignment='top', horizontalalignment='right',
                 bbox=dict(boxstyle='round,pad=0.3', fc='aliceblue', alpha=0.7))
        plt.tight_layout()
        filename = os.path.join(OUTPUT_DIR, f"dist_per_veh_{config_label}.png")
        plt.savefig(filename)
        chart_files['scenarios'][config_label]['distance_per_vehicle'] = filename
        plt.close(fig1)

        # Chart 1b: Stops per Vehicle (for this scenario)
        fig2, ax2 = plt.subplots(figsize=(max(6, num_vehicles*1.5), 4)) # Adjust width
        stops = [v["stops"] for v in vehicle_details]
        ax2.bar(v_ids, stops, color='orange')
        ax2.set_title(f"Stops per Vehicle - {config_label}")
        ax2.set_xlabel("Vehicle ID")
        ax2.set_ylabel("Number of Stops")
        ax2.text(0.98, 0.95, "Workload distribution by stops.",
                 transform=ax2.transAxes, fontsize=9, verticalalignment='top', horizontalalignment='right',
                 bbox=dict(boxstyle='round,pad=0.3', fc='lightyellow', alpha=0.7))
        plt.tight_layout()
        filename = os.path.join(OUTPUT_DIR, f"stops_per_veh_{config_label}.png")
        plt.savefig(filename)
        chart_files['scenarios'][config_label]['stops_per_vehicle'] = filename
        plt.close(fig2)

        # Chart 1c: Load Over Route Order (for this scenario)
        fig5, ax5 = plt.subplots(figsize=(10, 5))
        max_load_scenario = 0
        max_stops_scenario = 0
        colors = plt.cm.viridis(np.linspace(0, 1, num_vehicles))
        for i, vinfo in enumerate(vehicle_details):
            load_list = vinfo["load_list"]
            if len(load_list) > 1: # Only plot if route exists
                ax5.plot(range(len(load_list)), load_list, marker='.', linestyle='-', label=f"Vehicle {vinfo['vehicle_id']}", color=colors[i])
                if load_list:
                     max_load_scenario = max(max_load_scenario, max(load_list))
                max_stops_scenario = max(max_stops_scenario, len(load_list)-1)

        ax5.set_title(f"Load Profile - {config_label}")
        ax5.set_xlabel("Stop Index in Route (0=Depot Start)")
        ax5.set_ylabel("Passengers Onboard")
        ax5.legend(loc='upper left', bbox_to_anchor=(1, 1))
        ax5.grid(True, linestyle='--', alpha=0.6)
        ax5.set_xticks(np.arange(0, max_stops_scenario + 1, step=max(1, (max_stops_scenario+1)//10)))
        ax5.set_ylim(bottom=-0.5, top=max_load_scenario + 1)
        ax5.text(0.5, 1.02, "Tracks passenger count between stops for each vehicle.",
                 transform=ax5.transAxes, fontsize=9, verticalalignment='bottom', horizontalalignment='center',
                 bbox=dict(boxstyle='round,pad=0.3', fc='whitesmoke', alpha=0.8))
        plt.tight_layout()
        filename = os.path.join(OUTPUT_DIR, f"load_profile_{config_label}.png")
        plt.savefig(filename)
        chart_files['scenarios'][config_label]['load_over_route'] = filename
        plt.close(fig5)

        # Aggregate segment distances for the histogram
        all_segment_dists.extend(scenario_data["segment_distances"])

    # --- Generate Aggregate Charts ---
    print("\n--- Generating Aggregate Charts ---")

    # Agg Chart 1: Accessibility Demand vs. Supply (Grouped Bar) - Use data from ALL runs
    if scenario_labels_all:
        print("  Generating Accessibility Demand vs Supply chart...")
        num_scenarios_all = len(scenario_labels_all)
        x = np.arange(num_scenarios_all)
        width = 0.2
        fig3, ax3 = plt.subplots(figsize=(max(8, num_scenarios_all * 2), 5)) # Adjust width
        ax3.bar(x - 1.5*width, bike_requests_list_all, width, label='Bike Req.', color='darkgreen')
        ax3.bar(x - 0.5*width, wheelchair_requests_list_all, width, label='Wheelchair Req.', color='firebrick')
        ax3.bar(x + 0.5*width, bike_vehicles_list_all, width, label='Bike Veh.', color='skyblue')
        ax3.bar(x + 1.5*width, wheelchair_vehicles_list_all, width, label='Wheelchair Veh.', color='mediumpurple')

        ax3.set_ylabel("Count")
        ax3.set_title("Accessibility Demand vs. Supply (All Scenarios)")
        ax3.set_xticks(x)
        ax3.set_xticklabels(scenario_labels_all, rotation=20, ha='right')
        ax3.legend(loc='upper left', bbox_to_anchor=(1, 1))
        ax3.text(0.5, 1.02, "Compares requested vs. available accessibility features.",
                 transform=ax3.transAxes, fontsize=9, verticalalignment='bottom', horizontalalignment='center',
                 bbox=dict(boxstyle='round,pad=0.3', fc='whitesmoke', alpha=0.8))
        plt.ylim(0, max(max(bike_requests_list_all, default=0), max(wheelchair_requests_list_all, default=0),
                        max(bike_vehicles_list_all, default=0), max(wheelchair_vehicles_list_all, default=0)) * 1.1 + 1)
        plt.tight_layout()
        filename = os.path.join(OUTPUT_DIR, "agg_accessibility_demand_vs_supply.png")
        plt.savefig(filename)
        chart_files['aggregate']['accessibility_demand_vs_supply'] = filename
        plt.close(fig3)

    # Agg Chart 2: Distribution of Route Segment Distances (Histogram) - Use successful runs data
    if all_segment_dists:
        print("  Generating Segment Distance Distribution chart...")
        fig4, ax4 = plt.subplots(figsize=(8, 5))
        ax4.hist(all_segment_dists, bins=max(10, int(len(all_segment_dists)/15)), color='darkgrey', edgecolor='black')
        ax4.set_title("Distribution of Inter-Stop Distances (Successful Scenarios)")
        ax4.set_xlabel("Segment Distance (miles)")
        ax4.set_ylabel("Frequency")
        median_dist = np.median(all_segment_dists)
        mean_dist = np.mean(all_segment_dists)
        ax4.axvline(median_dist, color='red', linestyle='dashed', linewidth=1, label=f'Median: {median_dist:.1f} mi')
        ax4.axvline(mean_dist, color='blue', linestyle='dashed', linewidth=1, label=f'Mean: {mean_dist:.1f} mi')
        ax4.legend()
        ax4.text(0.5, 0.92,"Frequency of distances between consecutive stops.\nIndicates route sparsity.",
                 transform=ax4.transAxes, fontsize=9, verticalalignment='top', horizontalalignment='center',
                 bbox=dict(boxstyle='round,pad=0.3', fc='whitesmoke', alpha=0.8))
        plt.tight_layout()
        filename = os.path.join(OUTPUT_DIR, "agg_segment_distance_hist.png")
        plt.savefig(filename)
        chart_files['aggregate']['segment_distance_hist'] = filename
        plt.close(fig4)
    else:
         print("  Skipping segment distance histogram: No successful runs or no segments.")

    return chart_files


###############################################################################
# 7) MAIN: RUN CASE STUDIES, PLOT, & GENERATE STRUCTURED RESULTS
###############################################################################
def main():
    all_scenario_results = {} # Store results keyed by config_label
    # Add a place to store routes for successful runs
    successful_routes = {}
    vehicle_colors = ['#e41a1c', '#377eb8', '#4daf4a', '#984ea3', '#ff7f00', '#ffff33', '#a65628', '#f781bf', '#999999']

    # --- Run Simulations ---
    for config in TEST_CONFIGS:
        # Define config_label *first* using data from config dict
        num_requests = config["num_requests"]
        num_vehicles = config["num_vehicles"]
        region_size = config["region_size"]
        config_label = f"Req{num_requests}_Veh{num_vehicles}_Reg{region_size:.2f}" # MOVED DEFINITION HERE

        # Now you can use config_label
        print(f"[{datetime.datetime.now()}] Starting scenario: {config_label}")
        print(f"\n=== Running Scenario: {config.get('name', config_label)} ({config_label}) ===")

        # Create random data
        print(f"[{datetime.datetime.now()}] Data creating for {config_label}...") # Added print
        data, locations, passenger_node_to_name, accessibility_requirements, requests = \
            create_rural_data(num_requests=num_requests,
                              num_vehicles=num_vehicles,
                              region_size=region_size)
        print(f"[{datetime.datetime.now()}] Data created for {config_label}")

        # Calculate map center for this scenario
        avg_lat = np.mean([loc[1] for loc in locations if loc[0] != 'Depot'])
        avg_lon = np.mean([loc[2] for loc in locations if loc[0] != 'Depot'])
        map_center = [avg_lat, avg_lon] if pd.notna(avg_lat) else [locations[0][1], locations[0][2]]

        # Solve routing
        print(f"[{datetime.datetime.now()}] SOLVING {config_label}...")
        metrics, routes, solve_duration = solve_routing(data, locations, passenger_node_to_name,
                                                        accessibility_requirements,
                                                        ls_heuristic_str="GREEDY_DESCENT",
                                                        time_limit=TIME_LIMIT_SEC)
        print(f"[{datetime.datetime.now()}] SOLVED {config_label} - Status: {'Success' if metrics else 'Failed'}")


        # --- Store results ---
        if metrics is None:
             # Create failure metrics
             metrics = { # Fills in basic details needed for reports/analysis
                "distance_miles": np.nan, "sim_time_min": np.nan, "solve_duration_sec": solve_duration,
                "vehicle_details": [], "segment_distances": [], "status": "Failed",
                "bike_requests": sum(1 for r in requests if r["needs_bike"]),
                "wheelchair_requests": sum(1 for r in requests if r["needs_wheelchair"]),
                "bike_vehicles": sum(1 for v in data["vehicles"] if v["bike_capable"]),
                "wheelchair_vehicles": sum(1 for v in data["vehicles"] if v["wheelchair_accessible"]),
             }
        else:
             # Store routes if solve was successful
             successful_routes[config_label] = (routes, data, locations, passenger_node_to_name, accessibility_requirements, map_center) # Store needed info

        metrics.update({ # Add config details to metrics dict
            "Config": config_label, # Store the label used as key
            "NumRequests": num_requests,
            "NumVehicles": num_vehicles,
            "RegionSize": region_size
        })
        all_scenario_results[config_label] = metrics
        print(f"[{datetime.datetime.now()}] Stored results for {config_label}") # Added print

        # --- Optional: Add garbage collection after each scenario ---
        try: # Add try-except for gc import if needed
            import gc
            gc.collect()
            print(f"[{datetime.datetime.now()}] Ran garbage collection after {config_label}")
        except ImportError:
            pass # Ignore if gc module is not available


    # --- Post-Processing: Plots, Maps, Report ---

    print(f"\n[{datetime.datetime.now()}] Starting analyze_and_plot_structured...") # Added print
    chart_files = analyze_and_plot_structured(all_scenario_results)
    print(f"[{datetime.datetime.now()}] Finished analyze_and_plot_structured.") # Added print

    # Generate Maps (using stored routes, NO re-solving)
    print(f"\n[{datetime.datetime.now()}] Starting map generation...") # Added print
    for config_label, route_data in successful_routes.items():
         print(f"  [{datetime.datetime.now()}] Generating map for {config_label}")
         routes_for_map, data_map, loc_map, pntn_map, acc_req_map, center_map = route_data
         map_filename = os.path.join(OUTPUT_DIR, f"map_{config_label}.html")
         try:
             my_map = visualize_routes_clearer(routes_for_map, data_map, loc_map, pntn_map,
                                               acc_req_map, vehicle_colors, center_map)
             if my_map:
                 my_map.save(map_filename)
                 # Store map filename
                 if config_label not in chart_files['scenarios']: chart_files['scenarios'][config_label] = {}
                 chart_files['scenarios'][config_label]['map'] = map_filename
                 print(f"    Map saved: {map_filename}")
                 if IN_JUPYTER: display(my_map)
         except Exception as e:
             print(f"    Error generating map for {config_label}: {e}")
         print(f"  [{datetime.datetime.now()}] Finished map for {config_label}")


    # --- Generate Summary Table DataFrame ---
    print(f"\n[{datetime.datetime.now()}] Generating summary table...") # Added print
    summary_list = []
    for r in all_scenario_results.values():
         summary_list.append({
            "Scenario": r["Config"], # Use label
            "Status": r["status"],
            "Requests": r["NumRequests"],
            "Vehicles": r["NumVehicles"],
            "Region Size": f"{r['RegionSize']:.2f}",
            "Total Dist (mi)": f"{r['distance_miles']:.1f}" if pd.notna(r['distance_miles']) else "N/A",
            "Max Dur (min)": f"{r['sim_time_min']:.1f}" if pd.notna(r['sim_time_min']) else "N/A",
             # *** ADDED to show dropped requests in summary table ***
            "Dropped Req.": r.get('dropped_requests_count', 'N/A') if r['status'] == 'Success' else ('N/A' if r['status'] == 'Failed' else '?'),
            "Solve Time (s)": f"{r['solve_duration_sec']:.2f}"
        })
    summary_df = pd.DataFrame(summary_list)
    print("\n--- Aggregated Results Summary ---")
    print(summary_df.to_string(index=False))

    # --- Generate Overall Comparison Chart ---
    print(f"\n[{datetime.datetime.now()}] Generating overall comparison chart...") # Added print
    df_plot = pd.DataFrame([
        { "Config": r["Config"], "distance_miles": r["distance_miles"], "sim_time_min": r["sim_time_min"] }
         for r in all_scenario_results.values() if r['status'] == 'Success'
    ])

    comp_chart_filename = os.path.join(OUTPUT_DIR, "agg_comparison_summary.png")
    if not df_plot.empty:
        df_plot_indexed = df_plot.set_index("Config")
        fig_comp, axes_comp = plt.subplots(1, 2, figsize=(max(8, len(df_plot_indexed)*2), 5)) # Adjust width
        fig_comp.suptitle("Overall Performance Comparison (Successful Runs)")
        df_plot_indexed["distance_miles"].plot(kind="bar", ax=axes_comp[0], rot=30, color='navy', alpha=0.7)
        axes_comp[0].set_title("Total Distance")
        axes_comp[0].set_ylabel("Miles")
        axes_comp[0].set_xlabel("")
        df_plot_indexed["sim_time_min"].plot(kind="bar", ax=axes_comp[1], rot=30, color='maroon', alpha=0.7)
        axes_comp[1].set_title("Max Route Duration")
        axes_comp[1].set_ylabel("Minutes")
        axes_comp[1].set_xlabel("")
        plt.tight_layout(rect=[0, 0.03, 1, 0.93]) # Adjust for suptitle
        plt.savefig(comp_chart_filename)
        chart_files['aggregate']['comparison_summary'] = comp_chart_filename
        plt.close(fig_comp)
        print(f"Overall comparison chart saved to {comp_chart_filename}")
    else:
        print("No successful runs to generate overall comparison chart.")
        chart_files['aggregate']['comparison_summary'] = None


    # --- Build the structured HTML report ---
    print(f"\n[{datetime.datetime.now()}] Building HTML report...") # Added print
    # Pass the modified `all_scenario_results` which includes dropped counts
    build_structured_results_page(all_scenario_results, summary_df, chart_files)

    print("\nAnalysis complete.")


# Make sure the __main__ block only calls main()
if __name__ == "__main__":
    main()
```
